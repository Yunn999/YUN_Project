```python
# ============================================================
# MediaPipe Pose → 휴리스틱 라벨(프레임) → 슬라이딩 윈도우 → LSTM 분류
# - Kaggle IMVIA 폴더: 루트 바로 아래 6개 상위 폴더(스크린샷 구조 대응)
# - 샘플링: grab()/retrieve() 사용, fps_out으로 리샘플
# - 특징: (x,y,visibility) 33점 ⇒ 99차원/프레임 (0~1 범위 그대로)
# - 라벨: 휴리스틱 상태머신 (UPRIGHT=0 / FALLING,ON_GROUND=1)
# ============================================================
# pip install opencv-python mediapipe==0.10.15 numpy torch scikit-learn tqdm


# 중복 처리 => 재귀탐색, 경로 탐색
# 라벨 리라벨링 npz 숫자를 신뢰하지않고 경로기반 재매핑


# 슬라이딩 윈도우  이진 분류 
# 슬라이딩 윈도우 : 고정 길이의 프레임 묶음(윈도우)

# 몸의 각도(어깨,엉덩이의 벡터)를 통해 낙상을 구분하는 라벨을 만들고 슬라이딩 윈도우로 이진 분류  

import os, math, cv2, numpy as np
from dataclasses import dataclass
from typing import Tuple, List
from tqdm import tqdm
import mediapipe as mp
import torch, torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

# -----------------------------
# 0) 설정 (여기만 채워서 쓰세요)
# -----------------------------
@dataclass
class DS:
    # Kaggle 루트(스크린샷의 "낙상kaggle" 폴더)
    DATA_ROOT= r""      # 예: r"C:\Users\YUN\Documents\...\낙상kaggle"
    OUT_NPZ= "imvia_mppose_win.npz"
    # 루트 아래 상위 폴더들(존재하는 것만 사용)
    TOP_DIRS= ("Coffee_room_01","Coffee_room_02","Home_01","Home_02","Lecture_room","Office")


# LSTM 학습을 할 거면, 고정 길이 텐서가 필요해서 fps_out / win / stride 같은 세팅이 필수
@dataclass
class BuildCfg:
    fps_out= 24                              # 출력 샘플링 FPS (grab/retrieve로 리샘플)
    win = int(round(3.2 * fps_out))          # 윈도우 길이(프레임)
    stride = int(round(0.4 * fps_out))       # 윈도우 보폭(프레임)
    min_vis = 0.35                           # 가시성 품질 체크 임계
    save_npz = True                          # NPZ 저장 여부


@dataclass
class FallParams:         
    tilt_thr = 60.0
    hard_tilt = 65.0
    vel_fast_thr = 0.20
    dtilt_thr = 12.0
    height_thr = 0.85
    drop_thr = 0.20
    falling_max_sec = 2.0
    onground_hold_sec = 0.6

@dataclass
class TrainCfg:
    npz_path = DS.OUT_NPZ
    batch = 64
    epochs= 15
    lr = 1e-3
    hid = 128
    dropout = 0.2
    device = "cuda" if torch.cuda.is_available() else "cpu"

# -----------------------------
# 1) MediaPipe Pose
# -----------------------------
mp_pose = mp.solutions.pose
POSE = mp_pose.Pose(
    model_complexity=1,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

# Mediapipe 인덱스
IDX = dict(L_HIP=23, R_HIP=24, L_SHOULDER=11, R_SHOULDER=12, NOSE=0)

def get_landmarks_xyv(frame_bgr):
    """(33,3) = x,y,visibility (0~1). 실패 시 0."""
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    res = POSE.process(rgb)
    out = np.zeros((33, 3), dtype=np.float32)
    if res.pose_landmarks:
        for i, lm in enumerate(res.pose_landmarks.landmark):
            out[i, 0] = lm.x
            out[i, 1] = lm.y
            out[i, 2] = lm.visibility
    return out

def key_visibility_ok(lm, min_vis=0.35):
    pairs = [(IDX['L_SHOULDER'], IDX['R_SHOULDER']),
             (IDX['L_HIP'], IDX['R_HIP'])]
    for a, b in pairs:
        if lm[[a, b], 2].min() < min_vis:
            return False
    if lm[IDX['NOSE'], 2] < min_vis:
        return False
    return True

def _center_xy(landmarks):
    """어깨/엉덩이 중심 기준 평행이동(형상 각도 계산용)"""
    pts = landmarks[:, :2].copy()
    lhip, rhip = pts[IDX['L_HIP']], pts[IDX['R_HIP']]
    center = (lhip + rhip) / 2
    pts -= center
    return pts

def torso_tilt_deg(pts_centered):
    lsh, rsh = pts_centered[IDX['L_SHOULDER']], pts_centered[IDX['R_SHOULDER']]
    lhip, rhip = pts_centered[IDX['L_HIP']], pts_centered[IDX['R_HIP']]
    sh_c = (lsh + rsh) / 2
    hip_c = (lhip + rhip) / 2
    v = sh_c - hip_c
    ang = abs(math.degrees(math.atan2(v[0], -v[1])))  # 수직 기준; 서있음≈0°, 누울수록↑
    return float(ang)

def frame_label_heuristic(lm,
                          last_hip_y_raw,
                          fps_out,
                          params: FallParams,
                          last_tilt_s):
    """
    프레임 라벨링을 위해 필요한 값들 계산 + 상태 업데이트에 쓸 신호 반환.
    반환: (label(0/1), tilt_deg, dtilt_dps, head_h, hip_h, vel)
    """
    # (형상) 기울기
    pts_c = _center_xy(lm)
    tilt = torso_tilt_deg(pts_c)

    # (절대) 높이/속도
    head_h_raw = 1.0 - lm[IDX['NOSE'], 1]
    hip_y_raw = 0.5 * (lm[IDX['L_HIP'], 1] + lm[IDX['R_HIP'], 1])
    hip_h_raw = 1.0 - hip_y_raw

    # 엉덩이의 수직 속도. Δy × fps_out → 양수면 아래로 급강하.
    vel = 0.0 if last_hip_y_raw is None else (hip_y_raw - last_hip_y_raw) * fps_out

    # 간단 스무딩: 지수이동평균 한 스텝만(부담 줄이기)
    if last_tilt_s is None:
        tilt_s = tilt
    else:
        SMOOTH_WIN = 5
        k = 2 / (SMOOTH_WIN + 1)  # smooth_win=5 고정
        tilt_s = last_tilt_s * (1 - k) + tilt * k
    dtilt_s = 0.0 if last_tilt_s is None else (tilt_s - last_tilt_s) * fps_out

    # 임계 기반 라벨(프레임 단위; 간단 버전)

    # **‘자세가 낮다’**는 신호. 다음 중 하나라도 참이면 True
    # 기울기가 충분히 큼(tilt_s ≥ tilt_thr) // 머리 높이/엉덩이 높이가 낮음(≤ height_thr) // 머리 낙폭이 큼(1 - head_h_raw ≥ drop_thr)
    low_posture = ( (tilt_s >= params.tilt_thr) or (head_h_raw <= params.height_thr)
                    or (hip_h_raw <= params.height_thr) or ((1.0 - head_h_raw) >= params.drop_thr) )
    
    # 급격한 변화 신호:
    # 엉덩이의 하강 속도가 빠르다. / 기울기 변화속도가 빠르고(dtilt_s > dtilt_thr) 현재 기울기도 어느 정도 있다(tilt_s > 30) / 하드 기울기(매우 누움) 자체로도 급변으로 인정.
    fast_drop  = (vel > params.vel_fast_thr) or (dtilt_s > params.dtilt_thr and tilt_s > 30) or (tilt_s >= params.hard_tilt)

    label = 1 if (low_posture and fast_drop) or tilt_s >= params.hard_tilt else 0

    return label, tilt_s, dtilt_s, head_h_raw, hip_h_raw, vel


# -----------------------------
# 2) 비디오 스캔 및 시퀀스 생성
# -----------------------------
def _iter_scene_dirs(root ,top_dirs):
    for d in top_dirs:
        p = os.path.join(root, d)
        if os.path.isdir(p):    # 실제로 폴더가 존재할 때만
            yield d, p          # ("Home_01", "/root/Home_01") 같은 튜플을 생성

def _video_base_dir(scene_dir):
    """scene/Videos 또는 scene/*/Videos 자동 인식"""
    direct = os.path.join(scene_dir, "Videos")
    if os.path.isdir(direct):
        return direct
    
    try:
        for name in os.listdir(scene_dir):          # scene의 하위 폴더들을 뒤져서
            sub = os.path.join(scene_dir, name)
            cand = os.path.join(sub, "Videos")
            if os.path.isdir(cand):
                return cand
    except Exception:
        pass
    
    return scene_dir


def _iter_avi_recursive(base_dir):

        # r은 현재 디렉터리 경로, _는 (사용하지 않는) 하위 폴더 목록, files는 파일 목록
    for r, _, files in os.walk(base_dir):       # os.walk() => base_dir 아래의 모든 하위 폴더를 재귀적으로 순회
        for f in files:
            if f.lower().endswith(".avi"):
                yield os.path.join(r, f)



def sample_video_to_windows(video_path,
                            fps_out,
                            win,
                            stride,
                            min_vis,
                            params: FallParams):
    """
    비디오 1개 → (N,win,99), (N,)
    - grab/retrieve로 fps_out 샘플링
    - 프레임 입력: (x,y,vis)*33 → 99
    - 프레임 라벨: 휴리스틱(0/1)
    - 윈도우 라벨: 창 안에 1이 존재하면 1
    """

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64)

    fps_in = cap.get(cv2.CAP_PROP_FPS)
    step = max(1, int(round(fps_in / float(fps_out))))

    frames, labels = [], []
    idx = 0
    last_hip_y_raw = None
    last_tilt_s = None

    while True:
        ok = cap.grab()
        if not ok:
            break

        if idx % step == 0:
            ok, frame = cap.retrieve()
            if not ok:
                break

            lm = get_landmarks_xyv(frame)  # (33,3) 0~1 // 33개 관절의 (x, y, visibility)

            if not key_visibility_ok(lm, min_vis):
                # 시그널/특징 품질이 낮으면 제로
                frames.append(np.zeros((33,3), np.float32))
                labels.append(0)
            else:               # frame_label_heuristic가 기울기/변화속도/머리·엉덩이 높이/속도를 계산하고 프레임 라벨(0/1) 을 반환
                label, tilt_s, dtilt_s, head_h, hip_h, vel = frame_label_heuristic(
                    lm, last_hip_y_raw, fps_out, params, last_tilt_s
                )
                frames.append(lm)
                labels.append(label)
                last_hip_y_raw = 0.5 * (lm[IDX['L_HIP'], 1] + lm[IDX['R_HIP'], 1])
                last_tilt_s = tilt_s

        idx += 1    # 프레임 인덱스 증가 
    cap.release()


    if len(frames) < win:
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64)

    seq = np.stack(frames, axis=0).astype(np.float32)     # (T,33,3)
    X_t = seq.reshape(len(seq), 33*3)                     # (T,99)
    y_t = np.asarray(labels, dtype=np.int64)              # (T,)

    # 슬라이딩 윈도우
    chunks, wlabs = [], []
    for s in range(0, len(X_t) - win + 1, stride):
        e = s + win
        chunks.append(X_t[s:e])
        wlabs.append(1 if y_t[s:e].max() == 1 else 0)

    X = np.stack(chunks, axis=0) if chunks else np.zeros((0, win, 99), np.float32)
    y = np.asarray(wlabs, dtype=np.int64) if wlabs else np.zeros((0,), np.int64)
    return X, y



def build_npz_from_kaggle(ds: DS, b: BuildCfg, fp: FallParams):
    # 필수 설정 확인. 루트 경로가 비어 있으면 즉시 중단
    assert ds.DATA_ROOT, "DS.DATA_ROOT를 먼저 채워주세요."

    X_list, y_list, g_list = [], [], []

    for scene, scene_dir in _iter_scene_dirs(ds.DATA_ROOT, ds.TOP_DIRS):        # ("Home_01", "/root/Home_01") 같은 튜플
        base = _video_base_dir(scene_dir)   # video 폴더 
        vids = list(_iter_avi_recursive(base))  # .avi 동영상
        print(f"[SCAN] {scene}: {len(vids)}개 .avi")

        for v in tqdm(vids, desc=scene, unit="vid"):

            # Xv: (N_v, win, 99) (이 영상에서 나온 윈도우 수 × 윈도우 길이 × 99특징)
            # yv: (N_v,) (각 윈도우의 0/1 라벨)
            # 윈도우 : 연속된 프레임 win장을 붙여 만든 고정 길이 시퀀스
            # stride: 다음 윈도우를 만들 때 시작 프레임을 얼마나 건너뛸지
            Xv, yv = sample_video_to_windows(v, b.fps_out, b.win, b.stride, b.min_vis, fp)
            if len(Xv) == 0: 
                continue
            X_list.append(Xv)
            y_list.append(yv)
            g_list += [os.path.splitext(os.path.basename(v))[0]] * len(yv)
            # os.path.basename(v)는 파일명 // splitext(...)[0]는 확장자를 뗀 영상 이름

    if not X_list:
        raise RuntimeError("윈도우가 만들어지지 않았습니다.")

    X = np.concatenate(X_list, axis=0)  # (N,win,99)
    y = np.concatenate(y_list, axis=0)  # (N,)
    groups = np.array(g_list)

    print(f"[NPZ] X={X.shape}, y={y.shape}, groups={groups.shape}")
    if b.save_npz:
        np.savez_compressed(ds.OUT_NPZ, X=X, y=y, groups=groups,
                            info=dict(WIN=b.win, STRIDE=b.stride, FEAT="xyv(33*3)"))
        print(f"저장 완료 → {ds.OUT_NPZ}")
    return X, y, groups


from sklearn.metrics import confusion_matrix, classification_report   
import matplotlib.pyplot as plt                                       


# eval_confusion_matrix
def eval_confusion_matrix(model,
                          te_dl,
                          y_true,
                          device,
                          thr= 0.5,
                          save_png = "confusion_matrix_best.png"):
    """
    베스트 모델로 테스트 세트 예측 → 컨퓨전 매트릭스/리포트 출력 및(옵션) 저장
    """
    model.eval()
    probs = []
    with torch.no_grad():
        for xb, _ in te_dl:
            xb = xb.to(device)
            p = torch.sigmoid(model(xb)).cpu().numpy()
            probs.append(p)
    probs = np.concatenate(probs)
    y_pred = (probs >= thr).astype(np.int64)

    # 1) 텍스트 출력
    cm = confusion_matrix(y_true, y_pred)
    print("\n[Confusion Matrix] (rows=true, cols=pred)")
    print(cm)
    print("\n[Classification Report]")
    print(classification_report(y_true, y_pred, target_names=["UPRIGHT(0)", "FALL(1)"]))

    # 2) 시각화 (카운트 + 비율 함께 표기)
    cm_norm = confusion_matrix(y_true, y_pred, normalize='true')
    fig, ax = plt.subplots(figsize=(4.5, 4.5))
    im = ax.imshow(cm, cmap='Blues')
    ax.set_title("Confusion Matrix (best F1)")
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")
    ax.set_xticks([0, 1]); ax.set_yticks([0, 1])
    ax.set_xticklabels(["UPRIGHT(0)", "FALL(1)"])
    ax.set_yticklabels(["UPRIGHT(0)", "FALL(1)"])

    # 셀 안에 count와 비율(%) 표기
    for i in range(2):
        for j in range(2):
            ax.text(j, i, f"{cm[i, j]}\n({cm_norm[i, j]*100:.1f}%)",
                    ha="center", va="center", fontsize=10)

    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    plt.tight_layout()
    if save_png:
        plt.savefig(save_png, dpi=150, bbox_inches="tight")
        print(f"[Saved] {save_png}")
    plt.show()

    # 필요 시 개별 지표 반환
    return {
        "cm": cm,
        "cm_norm": cm_norm,
        "y_pred": y_pred,
        "probs": probs,
    }


# -----------------------------
# 3) LSTM 학습/평가
# -----------------------------
class WinDS(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X).float()
        self.y = torch.from_numpy(y).float()
    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]


class LSTMCls(nn.Module):
    def __init__(self, feat_dim=99, hid=128, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size=feat_dim, hidden_size=hid, num_layers=1,
                            batch_first=True, bidirectional=True)
        self.head = nn.Sequential(
            nn.LayerNorm(hid*2),
            nn.Dropout(dropout),
            nn.Linear(hid*2, hid),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hid, 1)
        )

        
    def forward(self, x):
        out, _ = self.lstm(x)          # (B,T,2H)
        logits = self.head(out[:, -1]) # 마지막 타임스텝
        return logits.squeeze(1)


def train_lstm_from_npz(cfg: TrainCfg):
    npz = np.load(cfg.npz_path, allow_pickle=True)
    X, y, groups = npz["X"], npz["y"], npz["groups"]
    print("[LOAD]", X.shape, y.shape, groups.shape)

    # 그룹 단위 셔플 분할기를 생성 //  한 번만 분할하며 검증 비율은 20%
    gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
    tr, te = next(gss.split(X, y, groups))
    X_tr, y_tr = X[tr], y[tr]
    X_te, y_te = X[te], y[te]

    pos = y_tr.sum(); neg = len(y_tr) - pos
    pos_w = neg / max(pos, 1)
    print(f"pos_w={pos_w:.2f} (pos={int(pos)}, neg={int(neg)})")

    tr_dl = DataLoader(WinDS(X_tr, y_tr), batch_size=cfg.batch, shuffle=True)
    te_dl = DataLoader(WinDS(X_te, y_te), batch_size=cfg.batch, shuffle=False)

    model = LSTMCls(feat_dim=X.shape[2], hid=cfg.hid, dropout=cfg.dropout).to(cfg.device)
    opt = torch.optim.AdamW(model.parameters(), lr=cfg.lr)
    crit = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([pos_w], device=cfg.device))

    best = (-1, None)
    for ep in range(1, cfg.epochs+1):
        model.train(); tr_loss = 0.0
        for xb, yb in tr_dl:    # (입력, 타깃) 반복
            xb, yb = xb.to(cfg.device), yb.to(cfg.device)
            opt.zero_grad()                     # 이전 스텝의 누적 기울기를 0으로 초기화
            logit = model(xb)                   # 순전파로 로짓(시그모이드 이전 점수)을 계산
            loss = crit(logit, yb)              # 예측 로짓과 타깃으로 손실을 계
            loss.backward(); opt.step()         # 역전파로 기울기를 계산하고, 옵티마이저로 파라미터를 한 스텝 갱신
            tr_loss += loss.item() * len(xb)    # 배치 손실을 샘플 수 가중 합으로 누적
        tr_loss /= len(tr_dl.dataset)           # 에포크 학습 손실의 샘플 단위 평균을 계산

        model.eval(); preds, probs = [], []
        with torch.no_grad():   # 파이토치의 자동미분을 비활성화하여 메모리·속도를 최적화
            for xb, _ in te_dl:
                xb = xb.to(cfg.device)
                p = torch.sigmoid(model(xb)).cpu().numpy()  # 시그모이드로 확률로 변환하고, CPU 넘파이 배열로 가져옴옴
                preds.append((p >= 0.5).astype(np.int64)); probs.append(p)  # 0.5 임계로 확률을 이진 라벨로 변환해 preds에, 원본 확률을 probs에 누적
        preds = np.concatenate(preds); probs = np.concatenate(probs)        
        acc = accuracy_score(y_te, preds); f1 = f1_score(y_te, preds)
        try: auc = roc_auc_score(y_te, probs)
        except: auc = float("nan")
        print(f"[Ep{ep}] loss={tr_loss:.4f} | acc={acc:.4f} f1={f1:.4f} auc={auc:.4f}")

        if f1 > best[0]:
            best = (f1, model.state_dict()) # F1이 최고치를 갱신하면 현재 모델 가중치를 베스트로 저장

        if best[1] is not None: 
            torch.save(best[1], "lstm_mppose_best.pt")
            print("베스트 저장 → lstm_mppose_best.pt")

            # === 추가: 베스트 모델 로드 후 컨퓨전 매트릭스/리포트 출력 ===
            
            model.load_state_dict(best[1])  # 메모리의 베스트 가중치를 모델에 로드
            
            _ = eval_confusion_matrix(model, te_dl, y_te, cfg.device, thr=0.5,
                                    save_png="confusion_matrix_best.png")


# -----------------------------
# 4) 실행
# -----------------------------
if __name__ == "__main__":
    ds = DS(DATA_ROOT=r"./")                # ★여기에 Kaggle 루트 입력
    b  = BuildCfg(fps_out=20, win=32, stride=16, min_vis=0.35, save_npz=True)
    hp = FallParams()                     # 휴리스틱 임계치 (필요 시 미세조정)

    # (1) 윈도우 데이터 생성(+저장)
    X, y, groups = build_npz_from_kaggle(ds, b, hp)

    # (2) LSTM 학습/평가
    train_lstm_from_npz(TrainCfg(npz_path=ds.OUT_NPZ))
```

```python
# ============================================================
# MediaPipe Pose → 휴리스틱 라벨(프레임) → 슬라이딩 윈도우 → LSTM 분류
# - Kaggle IMVIA 폴더: 루트 바로 아래 6개 상위 폴더(스크린샷 구조 대응)
# - 샘플링: grab()/retrieve() 사용, fps_out으로 리샘플
# - 특징: (x,y,visibility) 33점 ⇒ 99차원/프레임 (0~1 범위 그대로)
# - 라벨: 휴리스틱 상태머신 (UPRIGHT=0 / FALLING,ON_GROUND=1)
# ============================================================
# pip install opencv-python mediapipe==0.10.15 numpy torch scikit-learn tqdm



import os, math, cv2, numpy as np
from dataclasses import dataclass
from typing import Tuple, List
from tqdm import tqdm
import mediapipe as mp
import torch, torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

# -----------------------------
# 0) 설정 (여기만 채워서 쓰세요)
# -----------------------------
@dataclass
class DS:
    # Kaggle 루트(스크린샷의 "낙상kaggle" 폴더)
    DATA_ROOT= r""      # 예: r"C:\Users\YUN\Documents\...\낙상kaggle"
    OUT_NPZ= "imvia_mppose_win.npz"
    # 루트 아래 상위 폴더들(존재하는 것만 사용)
    TOP_DIRS= ("Coffee_room_01","Coffee_room_02","Home_01","Home_02","Lecture_room","Office")


# LSTM 학습을 할 거면, 고정 길이 텐서가 필요해서 fps_out / win / stride 같은 세팅이 필수
@dataclass
class BuildCfg:
    fps_out= 20            # 출력 샘플링 FPS (grab/retrieve로 리샘플)
    win = 32               # 윈도우 길이(프레임)
    stride = 16            # 윈도우 보폭(프레임)
    min_vis = 0.35         # 가시성 품질 체크 임계
    save_npz = True        # NPZ 저장 여부

@dataclass
class FallParams:         
    tilt_thr = 60.0
    hard_tilt = 65.0
    vel_fast_thr = 0.20
    dtilt_thr = 12.0
    height_thr = 0.85
    drop_thr = 0.20
    falling_max_sec = 2.0
    onground_hold_sec = 0.6

@dataclass
class TrainCfg:
    npz_path = DS.OUT_NPZ
    batch = 64
    epochs= 15
    lr = 1e-3
    hid = 128
    dropout = 0.2
    device = "cuda" if torch.cuda.is_available() else "cpu"

# -----------------------------
# 1) MediaPipe Pose
# -----------------------------
mp_pose = mp.solutions.pose
POSE = mp_pose.Pose(
    model_complexity=1,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

# Mediapipe 인덱스
IDX = dict(L_HIP=23, R_HIP=24, L_SHOULDER=11, R_SHOULDER=12, NOSE=0)

def get_landmarks_xyv(frame_bgr):
    """(33,3) = x,y,visibility (0~1). 실패 시 0."""
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    res = POSE.process(rgb)
    out = np.zeros((33, 3), dtype=np.float32)
    if res.pose_landmarks:
        for i, lm in enumerate(res.pose_landmarks.landmark):
            out[i, 0] = lm.x
            out[i, 1] = lm.y
            out[i, 2] = lm.visibility
    return out

def key_visibility_ok(lm, min_vis=0.35):
    pairs = [(IDX['L_SHOULDER'], IDX['R_SHOULDER']),
             (IDX['L_HIP'], IDX['R_HIP'])]
    for a, b in pairs:
        if lm[[a, b], 2].min() < min_vis:
            return False
    if lm[IDX['NOSE'], 2] < min_vis:
        return False
    return True

def _center_xy(landmarks):
    """어깨/엉덩이 중심 기준 평행이동(형상 각도 계산용)"""
    pts = landmarks[:, :2].copy()
    lhip, rhip = pts[IDX['L_HIP']], pts[IDX['R_HIP']]
    center = (lhip + rhip) / 2
    pts -= center
    return pts

def torso_tilt_deg(pts_centered):
    lsh, rsh = pts_centered[IDX['L_SHOULDER']], pts_centered[IDX['R_SHOULDER']]
    lhip, rhip = pts_centered[IDX['L_HIP']], pts_centered[IDX['R_HIP']]
    sh_c = (lsh + rsh) / 2
    hip_c = (lhip + rhip) / 2
    v = sh_c - hip_c
    ang = abs(math.degrees(math.atan2(v[0], -v[1])))  # 수직 기준; 서있음≈0°, 누울수록↑
    return float(ang)

def frame_label_heuristic(lm,
                          last_hip_y_raw,
                          fps_out,
                          params: FallParams,
                          last_tilt_s):
    """
    프레임 라벨링을 위해 필요한 값들 계산 + 상태 업데이트에 쓸 신호 반환.
    반환: (label(0/1), tilt_deg, dtilt_dps, head_h, hip_h, vel)
    """
    # (형상) 기울기
    pts_c = _center_xy(lm)
    tilt = torso_tilt_deg(pts_c)

    # (절대) 높이/속도
    head_h_raw = 1.0 - lm[IDX['NOSE'], 1]
    hip_y_raw = 0.5 * (lm[IDX['L_HIP'], 1] + lm[IDX['R_HIP'], 1])
    hip_h_raw = 1.0 - hip_y_raw

    # 엉덩이의 수직 속도. Δy × fps_out → 양수면 아래로 급강하.
    vel = 0.0 if last_hip_y_raw is None else (hip_y_raw - last_hip_y_raw) * fps_out

    # 간단 스무딩: 지수이동평균 한 스텝만(부담 줄이기)
    if last_tilt_s is None:
        tilt_s = tilt
    else:
        SMOOTH_WIN = 5
        k = 2 / (SMOOTH_WIN + 1)  # smooth_win=5 고정
        tilt_s = last_tilt_s * (1 - k) + tilt * k
    dtilt_s = 0.0 if last_tilt_s is None else (tilt_s - last_tilt_s) * fps_out

    # 임계 기반 라벨(프레임 단위; 간단 버전)

    # **‘자세가 낮다’**는 신호. 다음 중 하나라도 참이면 True
    # 기울기가 충분히 큼(tilt_s ≥ tilt_thr) // 머리 높이/엉덩이 높이가 낮음(≤ height_thr) // 머리 낙폭이 큼(1 - head_h_raw ≥ drop_thr)
    low_posture = ( (tilt_s >= params.tilt_thr) or (head_h_raw <= params.height_thr)
                    or (hip_h_raw <= params.height_thr) or ((1.0 - head_h_raw) >= params.drop_thr) )
    
    # 급격한 변화 신호:
    # 엉덩이의 하강 속도가 빠르다. / 기울기 변화속도가 빠르고(dtilt_s > dtilt_thr) 현재 기울기도 어느 정도 있다(tilt_s > 30) / 하드 기울기(매우 누움) 자체로도 급변으로 인정.
    fast_drop  = (vel > params.vel_fast_thr) or (dtilt_s > params.dtilt_thr and tilt_s > 30) or (tilt_s >= params.hard_tilt)

    label = 1 if (low_posture and fast_drop) or tilt_s >= params.hard_tilt else 0

    return label, tilt_s, dtilt_s, head_h_raw, hip_h_raw, vel


# -----------------------------
# 2) 비디오 스캔 및 시퀀스 생성
# -----------------------------
def _iter_scene_dirs(root ,top_dirs):
    for d in top_dirs:
        p = os.path.join(root, d)
        if os.path.isdir(p):    # 실제로 폴더가 존재할 때만
            yield d, p          # ("Home_01", "/root/Home_01") 같은 튜플을 생성

def _video_base_dir(scene_dir):
    """scene/Videos 또는 scene/*/Videos 자동 인식"""
    direct = os.path.join(scene_dir, "Videos")
    if os.path.isdir(direct):
        return direct
    
    try:
        for name in os.listdir(scene_dir):          # scene의 하위 폴더들을 뒤져서
            sub = os.path.join(scene_dir, name)
            cand = os.path.join(sub, "Videos")
            if os.path.isdir(cand):
                return cand
    except Exception:
        pass
    
    return scene_dir


def _iter_avi_recursive(base_dir):

        # r은 현재 디렉터리 경로, _는 (사용하지 않는) 하위 폴더 목록, files는 파일 목록
    for r, _, files in os.walk(base_dir):       # os.walk() => base_dir 아래의 모든 하위 폴더를 재귀적으로 순회
        for f in files:
            if f.lower().endswith(".avi"):
                yield os.path.join(r, f)



def sample_video_to_windows(video_path,
                            fps_out,
                            win,
                            stride,
                            min_vis,
                            params: FallParams):
    """
    비디오 1개 → (N,win,99), (N,)
    - grab/retrieve로 fps_out 샘플링
    - 프레임 입력: (x,y,vis)*33 → 99
    - 프레임 라벨: 휴리스틱(0/1)
    - 윈도우 라벨: 창 안에 1이 존재하면 1
    """

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64)

    fps_in = cap.get(cv2.CAP_PROP_FPS)
    step = max(1, int(round(fps_in / float(fps_out))))

    frames, labels = [], []
    idx = 0
    last_hip_y_raw = None
    last_tilt_s = None

    while True:
        ok = cap.grab()
        if not ok:
            break

        if idx % step == 0:
            ok, frame = cap.retrieve()
            if not ok:
                break

            lm = get_landmarks_xyv(frame)  # (33,3) 0~1 // 33개 관절의 (x, y, visibility)

            if not key_visibility_ok(lm, min_vis):
                # 시그널/특징 품질이 낮으면 제로
                frames.append(np.zeros((33,3), np.float32))
                labels.append(0)
            else:               # frame_label_heuristic가 기울기/변화속도/머리·엉덩이 높이/속도를 계산하고 프레임 라벨(0/1) 을 반환
                label, tilt_s, dtilt_s, head_h, hip_h, vel = frame_label_heuristic(
                    lm, last_hip_y_raw, fps_out, params, last_tilt_s
                )
                frames.append(lm)
                labels.append(label)
                last_hip_y_raw = 0.5 * (lm[IDX['L_HIP'], 1] + lm[IDX['R_HIP'], 1])
                last_tilt_s = tilt_s

        idx += 1    # 프레임 인덱스 증가 
    cap.release()


    if len(frames) < win:
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64)

    seq = np.stack(frames, axis=0).astype(np.float32)     # (T,33,3)
    X_t = seq.reshape(len(seq), 33*3)                     # (T,99)
    y_t = np.asarray(labels, dtype=np.int64)              # (T,)

    # 슬라이딩 윈도우
    chunks, wlabs = [], []
    for s in range(0, len(X_t) - win + 1, stride):
        e = s + win
        chunks.append(X_t[s:e])
        wlabs.append(1 if y_t[s:e].max() == 1 else 0)

    X = np.stack(chunks, axis=0) if chunks else np.zeros((0, win, 99), np.float32)
    y = np.asarray(wlabs, dtype=np.int64) if wlabs else np.zeros((0,), np.int64)
    return X, y



def build_npz_from_kaggle(ds: DS, b: BuildCfg, fp: FallParams):
    # 필수 설정 확인. 루트 경로가 비어 있으면 즉시 중단
    assert ds.DATA_ROOT, "DS.DATA_ROOT를 먼저 채워주세요."

    X_list, y_list, g_list = [], [], []

    for scene, scene_dir in _iter_scene_dirs(ds.DATA_ROOT, ds.TOP_DIRS):        # ("Home_01", "/root/Home_01") 같은 튜플
        base = _video_base_dir(scene_dir)   # video 폴더 
        vids = list(_iter_avi_recursive(base))  # .avi 동영상
        print(f"[SCAN] {scene}: {len(vids)}개 .avi")

        for v in tqdm(vids, desc=scene, unit="vid"):

            # Xv: (N_v, win, 99) (이 영상에서 나온 윈도우 수 × 윈도우 길이 × 99특징)
            # yv: (N_v,) (각 윈도우의 0/1 라벨)
            # 윈도우 : 연속된 프레임 win장을 붙여 만든 고정 길이 시퀀스
            # stride: 다음 윈도우를 만들 때 시작 프레임을 얼마나 건너뛸지
            Xv, yv = sample_video_to_windows(v, b.fps_out, b.win, b.stride, b.min_vis, fp)
            if len(Xv) == 0: 
                continue
            X_list.append(Xv)
            y_list.append(yv)
            g_list += [os.path.splitext(os.path.basename(v))[0]] * len(yv)
            # os.path.basename(v)는 파일명 // splitext(...)[0]는 확장자를 뗀 영상 이름

    if not X_list:
        raise RuntimeError("윈도우가 만들어지지 않았습니다.")

    X = np.concatenate(X_list, axis=0)  # (N,win,99)
    y = np.concatenate(y_list, axis=0)  # (N,)
    groups = np.array(g_list)

    print(f"[NPZ] X={X.shape}, y={y.shape}, groups={groups.shape}")
    if b.save_npz:
        np.savez_compressed(ds.OUT_NPZ, X=X, y=y, groups=groups,
                            info=dict(WIN=b.win, STRIDE=b.stride, FEAT="xyv(33*3)"))
        print(f"저장 완료 → {ds.OUT_NPZ}")
    return X, y, groups


from sklearn.metrics import confusion_matrix, classification_report   
import matplotlib.pyplot as plt                                       


# eval_confusion_matrix
def eval_confusion_matrix(model,
                          te_dl,
                          y_true,
                          device,
                          thr= 0.5,
                          save_png = "confusion_matrix_best.png"):
    """
    베스트 모델로 테스트 세트 예측 → 컨퓨전 매트릭스/리포트 출력 및(옵션) 저장
    """
    model.eval()
    probs = []
    with torch.no_grad():
        for xb, _ in te_dl:
            xb = xb.to(device)
            p = torch.sigmoid(model(xb)).cpu().numpy()
            probs.append(p)
    probs = np.concatenate(probs)
    y_pred = (probs >= thr).astype(np.int64)

    # 1) 텍스트 출력
    cm = confusion_matrix(y_true, y_pred)
    print("\n[Confusion Matrix] (rows=true, cols=pred)")
    print(cm)
    print("\n[Classification Report]")
    print(classification_report(y_true, y_pred, target_names=["UPRIGHT(0)", "FALL(1)"]))

    # 2) 시각화 (카운트 + 비율 함께 표기)
    cm_norm = confusion_matrix(y_true, y_pred, normalize='true')
    fig, ax = plt.subplots(figsize=(4.5, 4.5))
    im = ax.imshow(cm, cmap='Blues')
    ax.set_title("Confusion Matrix (best F1)")
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")
    ax.set_xticks([0, 1]); ax.set_yticks([0, 1])
    ax.set_xticklabels(["UPRIGHT(0)", "FALL(1)"])
    ax.set_yticklabels(["UPRIGHT(0)", "FALL(1)"])

    # 셀 안에 count와 비율(%) 표기
    for i in range(2):
        for j in range(2):
            ax.text(j, i, f"{cm[i, j]}\n({cm_norm[i, j]*100:.1f}%)",
                    ha="center", va="center", fontsize=10)

    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    plt.tight_layout()
    if save_png:
        plt.savefig(save_png, dpi=150, bbox_inches="tight")
        print(f"[Saved] {save_png}")
    plt.show()

    # 필요 시 개별 지표 반환
    return {
        "cm": cm,
        "cm_norm": cm_norm,
        "y_pred": y_pred,
        "probs": probs,
    }


# -----------------------------
# 3) LSTM 학습/평가
# -----------------------------
class WinDS(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X).float()
        self.y = torch.from_numpy(y).float()
    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]


class LSTMCls(nn.Module):
    def __init__(self, feat_dim=99, hid=128, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size=feat_dim, hidden_size=hid, num_layers=1,
                            batch_first=True, bidirectional=True)
        self.head = nn.Sequential(
            nn.LayerNorm(hid*2),
            nn.Dropout(dropout),
            nn.Linear(hid*2, hid),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hid, 1)
        )
    def forward(self, x):
        out, _ = self.lstm(x)          # (B,T,2H)
        logits = self.head(out[:, -1]) # 마지막 타임스텝
        return logits.squeeze(1)

def train_lstm_from_npz(cfg: TrainCfg):
    npz = np.load(cfg.npz_path, allow_pickle=True)
    X, y, groups = npz["X"], npz["y"], npz["groups"]
    print("[LOAD]", X.shape, y.shape, groups.shape)

    gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
    tr, te = next(gss.split(X, y, groups))
    X_tr, y_tr = X[tr], y[tr]
    X_te, y_te = X[te], y[te]

    pos = y_tr.sum(); neg = len(y_tr) - pos
    pos_w = neg / max(pos, 1)
    print(f"pos_w={pos_w:.2f} (pos={int(pos)}, neg={int(neg)})")

    tr_dl = DataLoader(WinDS(X_tr, y_tr), batch_size=cfg.batch, shuffle=True)
    te_dl = DataLoader(WinDS(X_te, y_te), batch_size=cfg.batch, shuffle=False)

    model = LSTMCls(feat_dim=X.shape[2], hid=cfg.hid, dropout=cfg.dropout).to(cfg.device)
    opt = torch.optim.AdamW(model.parameters(), lr=cfg.lr)
    crit = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([pos_w], device=cfg.device))

    best_f1 = -1.0
    best_state = None

    for ep in range(1, cfg.epochs+1):
        model.train(); tr_loss = 0.0
        for xb, yb in tr_dl:
            xb, yb = xb.to(cfg.device), yb.to(cfg.device)
            opt.zero_grad()
            logit = model(xb)
            loss = crit(logit, yb)
            loss.backward(); opt.step()
            tr_loss += loss.item() * len(xb)
        tr_loss /= len(tr_dl.dataset)

        # 검증 지표만 기록(여기서는 혼동행렬 계산 안 함)
        model.eval(); preds, probs = [], []
        with torch.no_grad():
            for xb, _ in te_dl:
                xb = xb.to(cfg.device)
                p = torch.sigmoid(model(xb)).cpu().numpy()
                preds.append((p >= 0.5).astype(np.int64)); probs.append(p)
        preds = np.concatenate(preds); probs = np.concatenate(probs)
        acc = accuracy_score(y_te, preds); f1 = f1_score(y_te, preds)
        try: auc = roc_auc_score(y_te, probs)
        except: auc = float("nan")
        print(f"[Ep{ep}] loss={tr_loss:.4f} | acc={acc:.4f} f1={f1:.4f} auc={auc:.4f}")

        if f1 > best_f1:
            best_f1 = f1
            best_state = model.state_dict()

    # === 학습 완료 후 '한 번만' 컨퓨전 매트릭스 계산 ===
    if best_state is not None:
        torch.save(best_state, "lstm_mppose_best.pt")
        print("베스트 저장 → lstm_mppose_best.pt")

        model.load_state_dict(best_state)
        model.eval()

        # 확률 → 예측
        probs = []
        with torch.no_grad():
            for xb, _ in te_dl:
                xb = xb.to(cfg.device)
                p = torch.sigmoid(model(xb)).cpu().numpy()
                probs.append(p)
        probs = np.concatenate(probs)
        y_pred = (probs >= 0.5).astype(np.int64)

        # 컨퓨전 매트릭스 및 리포트
        cm = confusion_matrix(y_te, y_pred)
        print("\n[Confusion Matrix] (rows=true, cols=pred)")
        print(cm)
        print("\n[Classification Report]")
        print(classification_report(y_te, y_pred, target_names=["UPRIGHT(0)", "FALL(1)"]))

        # 시각화 및 저장
        fig, ax = plt.subplots(figsize=(4.5, 4.5))
        im = ax.imshow(cm, cmap='Blues')
        ax.set_title("Confusion Matrix (best F1)")
        ax.set_xlabel("Predicted"); ax.set_ylabel("True")
        ax.set_xticks([0,1]); ax.set_yticks([0,1])
        ax.set_xticklabels(["UPRIGHT(0)", "FALL(1)"])
        ax.set_yticklabels(["UPRIGHT(0)", "FALL(1)"])
        for i in range(2):
            for j in range(2):
                ax.text(j, i, f"{cm[i, j]}", ha="center", va="center", fontsize=11)
        fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
        plt.tight_layout()
        plt.savefig("confusion_matrix_best.png", dpi=150, bbox_inches="tight")
        print("[Saved] confusion_matrix_best.png")
    else:
        print("베스트 가중치를 찾지 못했습니다.")


# -----------------------------
# 4) 실행
# -----------------------------
from types import SimpleNamespace

if __name__ == "__main__":
    ds = DS()
    ds.DATA_ROOT = r"./"  # 본인 경로로 수정

    # BuildCfg가 덮여있을 수 있으므로 SimpleNamespace 사용
    b  = SimpleNamespace(fps_out=20, win=32, stride=16, min_vis=0.35, save_npz=True)
    hp = FallParams()

    assert os.path.isdir(ds.DATA_ROOT), f"경로 없음: {ds.DATA_ROOT}"

    # (1) 윈도우 데이터 생성(+저장)
    X, y, groups = build_npz_from_kaggle(ds, b, hp)

    # TrainCfg도 덮여있을 수 있으므로 SimpleNamespace로 안전하게 전달
    cfg = SimpleNamespace(
        npz_path=ds.OUT_NPZ,
        batch=64,
        epochs=15,
        lr=1e-3,
        hid=128,
        dropout=0.2,
        device=("cuda" if torch.cuda.is_available() else "cpu"),
    )

    # (2) LSTM 학습/평가 (학습 종료 후 1회 컨퓨전 매트릭스 출력)
    train_lstm_from_npz(cfg)
```

```python
# ============================================================
# Fall detection end-to-end (copy-paste runnable)
# - Build windows with meta (video_path, start_frame_resampled)
# - Train BiLSTM once with group-safe split
# - Final eval: dump CSVs (all & errors) + confusion matrix PNG
# ------------------------------------------------------------
# pip install opencv-python mediapipe==0.10.15 numpy torch scikit-learn tqdm pandas matplotlib
# ============================================================

import os, math, cv2, numpy as np, pandas as pd, matplotlib.pyplot as plt
from dataclasses import dataclass
from typing import List, Tuple
from tqdm import tqdm
import mediapipe as mp
import torch, torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, classification_report

# -----------------------------
# 0) 설정
# -----------------------------
@dataclass
class DS:
    DATA_ROOT: str = r""     # ★★★ 여기를 본인 경로로 수정하세요 (예: r"D:\fall_kaggle")
    OUT_NPZ: str = "imvia_mppose_win.npz"
    TOP_DIRS: tuple = ("Coffee_room_01","Coffee_room_02","Home_01","Home_02","Lecture_room","Office")

@dataclass
class BuildCfg:
    fps_out: int = 20
    win: int = 32
    stride: int = 16
    min_vis: float = 0.35
    save_npz: bool = True

@dataclass
class FallParams:
    tilt_thr: float = 60.0
    hard_tilt: float = 65.0
    vel_fast_thr: float = 0.20
    dtilt_thr: float = 12.0
    height_thr: float = 0.85
    drop_thr: float = 0.20

@dataclass
class TrainCfg:
    npz_path: str
    batch: int = 64
    epochs: int = 15
    lr: float = 1e-3
    hid: int = 128
    dropout: float = 0.2
    device: str = "cuda" if torch.cuda.is_available() else "cpu"
    model_path: str = "lstm_mppose_best.pt"
    csv_all: str = "eval_windows_all.csv"
    csv_err: str = "eval_windows_errors.csv"
    cm_png: str = "confusion_matrix_best.png"

# -----------------------------
# 1) MediaPipe Pose
# -----------------------------
mp_pose = mp.solutions.pose
POSE = mp_pose.Pose(model_complexity=1, min_detection_confidence=0.5, min_tracking_confidence=0.5)

IDX = dict(L_HIP=23, R_HIP=24, L_SHOULDER=11, R_SHOULDER=12, NOSE=0)

def get_landmarks_xyv(frame_bgr: np.ndarray) -> np.ndarray:
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    res = POSE.process(rgb)
    out = np.zeros((33, 3), dtype=np.float32)
    if res.pose_landmarks:
        for i, lm in enumerate(res.pose_landmarks.landmark):
            out[i, 0] = lm.x; out[i, 1] = lm.y; out[i, 2] = lm.visibility
    return out

def key_visibility_ok(lm: np.ndarray, min_vis: float) -> bool:
    pairs = [(IDX['L_SHOULDER'], IDX['R_SHOULDER']), (IDX['L_HIP'], IDX['R_HIP'])]
    for a, b in pairs:
        if lm[[a, b], 2].min() < min_vis:
            return False
    if lm[IDX['NOSE'], 2] < min_vis:
        return False
    return True

def _center_xy(landmarks: np.ndarray) -> np.ndarray:
    pts = landmarks[:, :2].copy()
    lhip, rhip = pts[IDX['L_HIP']], pts[IDX['R_HIP']]
    center = (lhip + rhip) / 2
    pts -= center
    return pts

def torso_tilt_deg(pts_centered: np.ndarray) -> float:
    lsh, rsh = pts_centered[IDX['L_SHOULDER']], pts_centered[IDX['R_SHOULDER']]
    lhip, rhip = pts_centered[IDX['L_HIP']], pts_centered[IDX['R_HIP']]
    sh_c = (lsh + rsh) / 2
    hip_c = (lhip + rhip) / 2
    v = sh_c - hip_c
    ang = abs(math.degrees(math.atan2(v[0], -v[1])))
    return float(ang)

def frame_label_heuristic(lm: np.ndarray,
                          last_hip_y_raw: float,
                          fps_out: int,
                          params: FallParams,
                          last_tilt_s: float):
    # 기울기
    pts_c = _center_xy(lm)
    tilt = torso_tilt_deg(pts_c)
    # 높이/속도
    head_h_raw = 1.0 - lm[IDX['NOSE'], 1]
    hip_y_raw = 0.5 * (lm[IDX['L_HIP'], 1] + lm[IDX['R_HIP'], 1])
    hip_h_raw = 1.0 - hip_y_raw
    vel = 0.0 if last_hip_y_raw is None else (hip_y_raw - last_hip_y_raw) * fps_out
    # 간이 EMA 스무딩
    if last_tilt_s is None:
        tilt_s = tilt
    else:
        k = 2 / (5 + 1)
        tilt_s = last_tilt_s * (1 - k) + tilt * k
    dtilt_s = 0.0 if last_tilt_s is None else (tilt_s - last_tilt_s) * fps_out
    # 라벨
    low_posture = ((tilt_s >= params.tilt_thr) or (head_h_raw <= params.height_thr)
                   or (hip_h_raw <= params.height_thr) or ((1.0 - head_h_raw) >= params.drop_thr))
    fast_drop = (vel > params.vel_fast_thr) or (dtilt_s > params.dtilt_thr and tilt_s > 30) or (tilt_s >= params.hard_tilt)
    label = 1 if (low_posture and fast_drop) or tilt_s >= params.hard_tilt else 0
    return label, tilt_s, dtilt_s, head_h_raw, hip_h_raw, vel

# -----------------------------
# 2) 데이터 스캔 & 윈도우 생성
# -----------------------------
def _iter_scene_dirs(root: str, top_dirs: tuple):
    for d in top_dirs:
        p = os.path.join(root, d)
        if os.path.isdir(p):
            yield d, p

def _video_base_dir(scene_dir: str) -> str:
    direct = os.path.join(scene_dir, "Videos")
    if os.path.isdir(direct):
        return direct
    try:
        for name in os.listdir(scene_dir):
            sub = os.path.join(scene_dir, name)
            cand = os.path.join(sub, "Videos")
            if os.path.isdir(cand):
                return cand
    except Exception:
        pass
    return scene_dir

def _iter_avi_recursive(base_dir: str):
    for r, _, files in os.walk(base_dir):
        for f in files:
            if f.lower().endswith(".avi"):
                yield os.path.join(r, f)

def sample_video_to_windows(video_path: str,
                            fps_out: int,
                            win: int,
                            stride: int,
                            min_vis: float,
                            params: FallParams):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64), np.zeros((0,), np.int64)
    fps_in = cap.get(cv2.CAP_PROP_FPS)
    fps_in = fps_in if fps_in and fps_in > 0 else fps_out
    step = max(1, int(round(fps_in / float(fps_out))))

    frames, labels, starts = [], [], []
    idx = 0
    last_hip_y_raw = None
    last_tilt_s = None

    while True:
        ok = cap.grab()
        if not ok:
            break
        if idx % step == 0:
            ok, frame = cap.retrieve()
            if not ok:
                break
            lm = get_landmarks_xyv(frame)
            if not key_visibility_ok(lm, min_vis):
                frames.append(np.zeros((33,3), np.float32))
                labels.append(0)
            else:
                label, tilt_s, dtilt_s, head_h, hip_h, vel = frame_label_heuristic(
                    lm, last_hip_y_raw, fps_out, params, last_tilt_s)
                frames.append(lm); labels.append(label)
                last_hip_y_raw = 0.5 * (lm[IDX['L_HIP'], 1] + lm[IDX['R_HIP'], 1])
                last_tilt_s = tilt_s
        idx += 1
    cap.release()

    if len(frames) < win:
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64), np.zeros((0,), np.int64)

    seq = np.stack(frames, axis=0).astype(np.float32)    # (T,33,3)
    X_t = seq.reshape(len(seq), 99)                      # (T,99)
    y_t = np.asarray(labels, dtype=np.int64)             # (T,)

    chunks, wlabs = [], []
    for s in range(0, len(X_t) - win + 1, stride):
        e = s + win
        chunks.append(X_t[s:e])
        wlabs.append(1 if y_t[s:e].max() == 1 else 0)
        starts.append(s)
    X = np.stack(chunks, axis=0) if chunks else np.zeros((0, win, 99), np.float32)
    y = np.asarray(wlabs, dtype=np.int64) if wlabs else np.zeros((0,), np.int64)
    starts = np.asarray(starts, dtype=np.int64) if starts else np.zeros((0,), np.int64)
    return X, y, starts

def build_npz_from_kaggle(ds: DS, b: BuildCfg, fp: FallParams):
    assert ds.DATA_ROOT, "★ DS.DATA_ROOT를 먼저 채워주세요."
    assert os.path.isdir(ds.DATA_ROOT), f"경로 없음: {ds.DATA_ROOT}"

    X_list, y_list, g_list = [], [], []
    meta_vids, meta_starts = [], []

    for scene, scene_dir in _iter_scene_dirs(ds.DATA_ROOT, ds.TOP_DIRS):
        base = _video_base_dir(scene_dir)
        vids = list(_iter_avi_recursive(base))
        print(f"[SCAN] {scene}: {len(vids)}개 .avi")
        for v in tqdm(vids, desc=scene, unit="vid"):
            Xv, yv, sv = sample_video_to_windows(v, b.fps_out, b.win, b.stride, b.min_vis, fp)
            if len(Xv) == 0: 
                continue
            X_list.append(Xv); y_list.append(yv)
            g_list += [os.path.splitext(os.path.basename(v))[0]] * len(yv)
            meta_vids += [v] * len(sv)
            meta_starts.append(sv)

    if not X_list:
        raise RuntimeError("윈도우가 만들어지지 않았습니다.")

    X = np.concatenate(X_list, axis=0)
    y = np.concatenate(y_list, axis=0)
    groups = np.array(g_list)
    meta_start = np.concatenate(meta_starts, axis=0)
    meta_vid = np.array(meta_vids, dtype=object)

    print(f"[NPZ] X={X.shape}, y={y.shape}, groups={groups.shape}")
    if b.save_npz:
        np.savez_compressed(
            ds.OUT_NPZ,
            X=X, y=y, groups=groups,
            meta_vid=meta_vid,
            meta_start=meta_start,
            info=dict(WIN=b.win, STRIDE=b.stride, FEAT="xyv(33*3)", FPS_OUT=b.fps_out)
        )
        print(f"[Saved] {ds.OUT_NPZ}")
    return X, y, groups

# -----------------------------
# 3) Dataset / Model
# -----------------------------
class WinDS(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X).float()
        self.y = torch.from_numpy(y).float()
    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]

class LSTMCls(nn.Module):
    def __init__(self, feat_dim=99, hid=128, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size=feat_dim, hidden_size=hid, num_layers=1,
                            batch_first=True, bidirectional=True)
        self.head = nn.Sequential(
            nn.LayerNorm(hid*2), nn.Dropout(dropout),
            nn.Linear(hid*2, hid), nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(hid, 1)
        )
    def forward(self, x):
        out, _ = self.lstm(x)          # (B,T,2H)
        logits = self.head(out[:, -1]) # last timestep
        return logits.squeeze(1)

# -----------------------------
# 4) Confusion Matrix helper
# -----------------------------
def save_confusion_png(y_true, y_pred, path="confusion_matrix_best.png"):
    cm = confusion_matrix(y_true, y_pred)
    cm_norm = confusion_matrix(y_true, y_pred, normalize='true')
    fig, ax = plt.subplots(figsize=(4.5, 4.5))
    im = ax.imshow(cm, cmap='Blues')
    ax.set_title("Confusion Matrix (best F1)")
    ax.set_xlabel("Predicted"); ax.set_ylabel("True")
    ax.set_xticks([0, 1]); ax.set_yticks([0, 1])
    ax.set_xticklabels(["UPRIGHT(0)", "FALL(1)"]); ax.set_yticklabels(["UPRIGHT(0)", "FALL(1)"])
    for i in range(2):
        for j in range(2):
            ax.text(j, i, f"{cm[i, j]}\n({cm_norm[i, j]*100:.1f}%)", ha="center", va="center", fontsize=10)
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches="tight")
    plt.close(fig)
    print(f"[Saved] {path}")

# -----------------------------
# 5) Train + Final Eval (CSV/PNG)
# -----------------------------
def train_and_eval(cfg: TrainCfg):
    npz = np.load(cfg.npz_path, allow_pickle=True)
    X, y, groups = npz["X"], npz["y"], npz["groups"]
    meta_vid = npz["meta_vid"]; meta_start = npz["meta_start"]
    info = npz["info"].item() if "info" in npz.files else {}
    fps_out = info.get("FPS_OUT", 20); win = info.get("WIN", X.shape[1])

    print("[LOAD]", X.shape, y.shape, groups.shape)

    gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
    tr_idx, te_idx = next(gss.split(X, y, groups))
    X_tr, y_tr = X[tr_idx], y[tr_idx]
    X_te, y_te = X[te_idx], y[te_idx]
    meta_vid_te, meta_start_te = meta_vid[te_idx], meta_start[te_idx]

    pos = y_tr.sum(); neg = len(y_tr) - pos
    pos_w = float(neg) / max(float(pos), 1.0)
    print(f"pos_w={pos_w:.2f} (pos={int(pos)}, neg={int(neg)})")

    tr_dl = DataLoader(WinDS(X_tr, y_tr), batch_size=cfg.batch, shuffle=True)
    te_dl = DataLoader(WinDS(X_te, y_te), batch_size=cfg.batch, shuffle=False)

    model = LSTMCls(feat_dim=X.shape[2], hid=cfg.hid, dropout=cfg.dropout).to(cfg.device)
    opt = torch.optim.AdamW(model.parameters(), lr=cfg.lr)
    crit = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([pos_w], device=cfg.device))

    best_f1, best_sd = -1.0, None
    for ep in range(1, cfg.epochs + 1):
        # train
        model.train(); tr_loss = 0.0
        for xb, yb in tr_dl:
            xb, yb = xb.to(cfg.device), yb.to(cfg.device)
            opt.zero_grad()
            logit = model(xb)
            loss = crit(logit, yb)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()
            tr_loss += loss.item() * len(xb)
        tr_loss /= len(tr_dl.dataset)

        # val (light)
        model.eval(); probs, preds = [], []
        with torch.no_grad():
            for xb, _ in te_dl:
                xb = xb.to(cfg.device)
                p = torch.sigmoid(model(xb)).squeeze(-1).cpu().numpy()
                probs.append(p); preds.append((p >= 0.5).astype(np.int64))
        probs = np.concatenate(probs); preds = np.concatenate(preds)
        acc = accuracy_score(y_te, preds); f1 = f1_score(y_te, preds)
        try: auc = roc_auc_score(y_te, probs)
        except: auc = float("nan")
        print(f"[Ep{ep}] loss={tr_loss:.4f} | acc={acc:.4f} f1={f1:.4f} auc={auc:.4f}")

        if f1 > best_f1:
            best_f1 = f1; best_sd = model.state_dict().copy()

    # ---- Final eval with BEST ----
    print(f"[FINAL] Best F1 = {best_f1:.4f}")
    model.load_state_dict(best_sd)
    torch.save(best_sd, cfg.model_path); print(f"[Saved] {cfg.model_path}")

    model.eval(); probs, preds = [], []
    with torch.no_grad():
        for xb, _ in te_dl:
            xb = xb.to(cfg.device)
            p = torch.sigmoid(model(xb)).squeeze(-1).cpu().numpy()
            probs.append(p); preds.append((p >= 0.5).astype(np.int64))
    probs = np.concatenate(probs); preds = np.concatenate(preds)
    acc = accuracy_score(y_te, preds); f1 = f1_score(y_te, preds)
    try: auc = roc_auc_score(y_te, probs)
    except: auc = float("nan")
    print(f"[BEST] acc={acc:.4f} f1={f1:.4f} auc={auc:.4f}  (N={len(y_te)})")
    print("\n[Classification Report]\n" + classification_report(y_te, preds, target_names=["UPRIGHT(0)","FALL(1)"]))

    # ---- CSVs ----
    df_all = pd.DataFrame({
        "video_path": meta_vid_te,
        "start_frame_resampled": meta_start_te,
        "start_time_sec": meta_start_te / float(fps_out),
        "end_time_sec": (meta_start_te + int(win)) / float(fps_out),
        "y_true": y_te.astype(int),
        "y_pred": preds.astype(int),
        "prob_fall": probs.astype(float)
    })
    err_df = df_all[df_all["y_true"] != df_all["y_pred"]].reset_index(drop=True)
    df_all.to_csv(cfg.csv_all, index=False, encoding="utf-8-sig")
    err_df.to_csv(cfg.csv_err, index=False, encoding="utf-8-sig")
    print(f"[Saved] {cfg.csv_all} ({len(df_all)} rows)")
    print(f"[Saved] {cfg.csv_err} ({len(err_df)} wrong)")

    # ---- Confusion PNG ----
    save_confusion_png(y_te, preds, cfg.cm_png)

# -----------------------------
# 6) 실행
# -----------------------------
if __name__ == "__main__":
    ds = DS()
    ds.DATA_ROOT = r"./"  # ★★★ 꼭 채우세요. 예: r"D:\fall_kaggle"
    b = BuildCfg()
    fp = FallParams()

    # (1) 윈도우 생성(+NPZ 저장)
    X, y, groups = build_npz_from_kaggle(ds, b, fp)

    # (2) 학습 + 최종 평가/저장
    cfg = TrainCfg(npz_path=ds.OUT_NPZ)
    train_and_eval(cfg)
```

```python
import os
import pandas as pd

base_dir = "./라벨 구분/"

video_paths, labels, unique_names = [], [], []

for scene in os.listdir(base_dir):
    scene_path = os.path.join(base_dir, scene)
    if not os.path.isdir(scene_path):
        continue

    for label_name in ["fall", "non_fall"]:
        label_path = os.path.join(scene_path, label_name)
        if not os.path.isdir(label_path):
            continue

        label_value = 1 if label_name == "fall" else 0

        for filename in os.listdir(label_path):
            if filename.lower().endswith((".mp4", ".avi", ".mov", ".mkv")):
                file_path = os.path.join(label_path, filename)
                clean_name = filename.replace("(", "").replace(")", "")
                unique_name = f"{scene}_{label_name}_{clean_name}"

                video_paths.append(file_path)
                labels.append(label_value)
                unique_names.append(unique_name)

df = pd.DataFrame({
    "video_path": video_paths,
    "unique_name": unique_names,
    "label": labels
})

csv_path = "./video_dataset_index.csv"
df.to_csv(csv_path, index=False)
```

```python
import cv2
import numpy as np
import mediapipe as mp

mp_pose = mp.solutions.pose

def extract_keypoints(video_path, max_frames=60):
    cap = cv2.VideoCapture(video_path)
    pose = mp_pose.Pose(static_image_mode=False,
                        min_detection_confidence=0.5,
                        min_tracking_confidence=0.5)
    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_indices = np.linspace(0, total_frames - 1, max_frames).astype(int)

    for i in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, i)
        ret, frame = cap.read()
        if not ret:
            continue

        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        result = pose.process(frame_rgb)

        if result.pose_landmarks:
            lm = np.array([[p.x, p.y, p.visibility] for p in result.pose_landmarks.landmark])
        else:
            lm = np.zeros((33, 3))

        frames.append(lm.flatten())

    cap.release()
    pose.close()

    if len(frames) < max_frames:
        pad = np.zeros((max_frames - len(frames), 33 * 3))
        frames.extend(pad)

    return np.array(frames[:max_frames])
```

```python
from tqdm import tqdm
import numpy as np

X_data = []
for path in tqdm(df["video_path"]):
    X_data.append(extract_keypoints(path, max_frames=60))

X = np.array(X_data)
y = df["label"].values
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential([
    LSTM(128, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True),
    Dropout(0.6),
    LSTM(64),
    Dropout(0.6),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(
    X_train, y_train,
    epochs=25,
    batch_size=8,
    validation_split=0.2,
    verbose=1
)

loss, acc = model.evaluate(X_test, y_test)
print(f"✅ 테스트 정확도: {acc:.2f}")
```

```python
# ============================================================
# Sliding-window LSTM for Fall/Non-Fall (copy-paste runnable)
# - 각 비디오를 fps_out으로 리샘플 → (T, 99) 특징열
# - 길이 win, 보폭 stride로 슬라이딩 윈도우 생성
# - 라벨: 비디오 레이블을 윈도우에 상속(약지도)  ※ 프레임 GT 없으므로
# - Train/Test는 "비디오 단위"로 분할하여 누수 방지
# - 평가/혼동행렬 + 오분류 CSV + (옵션) 알림 트리거 데모
# ============================================================

import os, cv2, math, numpy as np, pandas as pd, matplotlib.pyplot as plt
from tqdm import tqdm
import mediapipe as mp

# -----------------------------
# 0) 설정
# -----------------------------
DATASET_CSV = "./video_dataset_index.csv"  # 사용자가 이미 만든 CSV: [video_path, unique_name, label]
FPS_OUT = 20
WIN = 32          # 윈도우 길이(프레임)
STRIDE = 16       # 윈도우 보폭(프레임)
MIN_VIS = 0.35    # 포즈 가시성 품질 임계
BATCH = 32
EPOCHS = 20
THR = 0.5         # 분류 임계(확률)

# 결과물 파일
CSV_ALL = "eval_windows_all.csv"
CSV_ERR = "eval_windows_errors.csv"

# -----------------------------
# 1) MediaPipe Pose: 프레임 → (33,3)
# -----------------------------
mp_pose = mp.solutions.pose
POSE = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)
IDX = dict(L_HIP=23, R_HIP=24, L_SHOULDER=11, R_SHOULDER=12, NOSE=0)

def key_visibility_ok(lm, min_vis=MIN_VIS):
    # 어깨·엉덩이·코 가시성 체크
    pairs = [(IDX['L_SHOULDER'], IDX['R_SHOULDER']),
             (IDX['L_HIP'], IDX['R_HIP'])]
    for a, b in pairs:
        if lm[[a, b], 2].min() < min_vis:
            return False
    if lm[IDX['NOSE'], 2] < min_vis:
        return False
    return True

def frame_to_xyv(frame_bgr):
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    res = POSE.process(rgb)
    out = np.zeros((33, 3), dtype=np.float32)
    if res.pose_landmarks:
        for i, lm in enumerate(res.pose_landmarks.landmark):
            out[i, 0] = lm.x; out[i, 1] = lm.y; out[i, 2] = lm.visibility
    return out

# -----------------------------
# 2) 비디오 → (T,99) 특징열 (fps_out으로 리샘플)
# -----------------------------
def video_to_sequence(video_path, fps_out=FPS_OUT):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return np.zeros((0, 99), np.float32)

    fps_in = cap.get(cv2.CAP_PROP_FPS)
    fps_in = fps_in if fps_in and fps_in > 0 else fps_out
    step = max(1, int(round(fps_in / float(fps_out))))

    feats = []
    idx = 0
    while True:
        ok = cap.grab()
        if not ok: break
        if idx % step == 0:
            ok, frame = cap.retrieve()
            if not ok: break
            lm = frame_to_xyv(frame)  # (33,3)
            if not key_visibility_ok(lm, MIN_VIS):
                lm[:] = 0.0
            feats.append(lm.reshape(-1))  # 99
        idx += 1
    cap.release()
    return np.asarray(feats, dtype=np.float32)  # (T,99)

# -----------------------------
# 3) 슬라이딩 윈도우 생성 (약지도: 비디오 레이블 상속)
# -----------------------------
def make_windows(seq_99, win=WIN, stride=STRIDE, label=0):
    """seq_99: (T,99) → X_win: (N,win,99), y_win: (N,) (모두 같은 label) + 시작프레임(리샘플 기준)"""
    T = len(seq_99)
    if T < win:
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64), np.zeros((0,), np.int64)
    chunks, labs, starts = [], [], []
    for s in range(0, T - win + 1, stride):
        e = s + win
        chunks.append(seq_99[s:e])
        labs.append(label)
        starts.append(s)
    X = np.stack(chunks, axis=0).astype(np.float32)
    y = np.asarray(labs, dtype=np.int64)
    starts = np.asarray(starts, dtype=np.int64)
    return X, y, starts

# -----------------------------
# 4) Dataset 빌드(Train/Test는 비디오 단위로 분할)
# -----------------------------
from sklearn.model_selection import train_test_split

def build_window_dataset(df, fps_out=FPS_OUT, win=WIN, stride=STRIDE):
    # 비디오 단위로 split
    tr_df, te_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])
    def build_from_df(dfx):
        X_list, y_list, meta_path, meta_start = [], [], [], []
        for _, row in tqdm(dfx.iterrows(), total=len(dfx), desc="Extract", unit="vid"):
            seq = video_to_sequence(row["video_path"], fps_out=fps_out)  # (T,99)
            Xw, yw, sw = make_windows(seq, win=win, stride=stride, label=int(row["label"]))
            if len(Xw)==0: continue
            X_list.append(Xw); y_list.append(yw)
            meta_path += [row["video_path"]]*len(sw)
            meta_start.append(sw)
        if not X_list:
            return (np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64),
                    np.array([], dtype=object), np.array([], dtype=np.int64))
        X = np.concatenate(X_list, axis=0)
        y = np.concatenate(y_list, axis=0)
        meta_start = np.concatenate(meta_start, axis=0)
        meta_path = np.array(meta_path, dtype=object)
        return X, y, meta_path, meta_start

    X_tr, y_tr, path_tr, start_tr = build_from_df(tr_df)
    X_te, y_te, path_te, start_te = build_from_df(te_df)
    info = dict(FPS_OUT=fps_out, WIN=win, STRIDE=stride)
    return (X_tr, y_tr, path_tr, start_tr), (X_te, y_te, path_te, start_te), info

# -----------------------------
# 5) Keras LSTM 모델
# -----------------------------
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, LayerNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

def build_model(input_shape, hid=128, drop=0.5):
    model = Sequential([
        LSTM(hid, input_shape=input_shape, return_sequences=True),
        Dropout(drop),
        LSTM(hid//2),
        Dropout(drop),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# -----------------------------
# 6) 실행
# -----------------------------
df = pd.read_csv(DATASET_CSV)
# 라벨이 혹시 문자열이면 매핑
if not np.issubdtype(df["label"].dtype, np.integer):
    df["label"] = df["label"].map({"non_fall":0, "Non-Fall":0, "fall":1, "Fall":1}).astype(int)

print(f"[INDEX] videos={len(df)}  pos={df['label'].sum()}  neg={len(df)-df['label'].sum()}")

# 데이터셋 만들기
(TR) , (TE), info = build_window_dataset(df, fps_out=FPS_OUT, win=WIN, stride=STRIDE)
X_tr, y_tr, path_tr, start_tr = TR
X_te, y_te, path_te, start_te = TE
print(f"[DATA] Train: X={X_tr.shape}, y={y_tr.shape} | Test: X={X_te.shape}, y={y_te.shape}")

# 모델 학습
model = build_model(input_shape=(X_tr.shape[1], X_tr.shape[2]), hid=128, drop=0.5)
cb = [
    EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", patience=2, factor=0.5, verbose=1),
]
history = model.fit(
    X_tr, y_tr, epochs=EPOCHS, batch_size=BATCH,
    validation_split=0.2, shuffle=True, callbacks=cb, verbose=1
)

# 평가
loss, acc = model.evaluate(X_te, y_te, verbose=0)
print(f"[TEST] loss={loss:.4f}  acc={acc:.4f}")

probs = model.predict(X_te, verbose=0).ravel()
y_pred = (probs >= THR).astype(int)

from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
try: auc = roc_auc_score(y_te, probs)
except: auc = float("nan")
print(f"[TEST] acc={accuracy_score(y_te, y_pred):.4f}  f1={f1_score(y_te, y_pred):.4f}  auc={auc:.4f}")
print("\n[Classification Report]\n" + classification_report(y_te, y_pred, target_names=["Non-Fall","Fall"]))

# 혼동행렬
cm = confusion_matrix(y_te, y_pred, labels=[0,1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non-Fall", "Fall"])
disp.plot(cmap="Blues", values_format='d')
plt.title("Confusion Matrix (Sliding-window LSTM)")
plt.tight_layout(); plt.show()

# 오분류 CSV (윈도우 단위)
df_all = pd.DataFrame({
    "video_path": path_te,
    "start_frame_resampled": start_te,
    "start_time_sec": start_te / float(info["FPS_OUT"]),
    "end_time_sec": (start_te + info["WIN"]) / float(info["FPS_OUT"]),
    "y_true": y_te.astype(int),
    "y_pred": y_pred.astype(int),
    "prob_fall": probs.astype(float),
})
err_df = df_all[df_all["y_true"] != df_all["y_pred"]].reset_index(drop=True)
df_all.to_csv(CSV_ALL, index=False, encoding="utf-8-sig")
err_df.to_csv(CSV_ERR, index=False, encoding="utf-8-sig")
print(f"[Saved] {CSV_ALL} ({len(df_all)})  |  {CSV_ERR} ({len(err_df)})")

# -----------------------------
# 7) (선택) 알림 트리거 데모
# -----------------------------
def trigger_alert_on_video(video_path, model, fps_out=FPS_OUT, win=WIN, stride=1, thr=THR, min_votes=3):
    """
    실시간/오프라인 공용 데모:
    - 한 프레임씩 새로 들어온다고 가정하고 슬라이딩 버퍼를 유지
    - 윈도우 확률이 thr 이상으로 '연속 min_votes번' 나오면 알림
    """
    seq = video_to_sequence(video_path, fps_out=fps_out)  # (T,99)
    if len(seq) < win: 
        print("[ALERT] too short"); return []
    alerts = []
    buf = []
    for t in range(len(seq)):
        buf.append(seq[t])
        if len(buf) > win: buf.pop(0)
        if len(buf) == win and (t % stride == 0):
            x = np.asarray(buf, dtype=np.float32)[None, ...]  # (1,win,99)
            p = float(model.predict(x, verbose=0).ravel()[0])
            # 연속 투표(간단)
            if p >= thr:
                alerts.append((t - win + 1, t, p))
            # 실제 실시간이면 여기서 즉시 push/사이렌 등 처리
    # 후처리: 연속 min_votes 이상만 유지
    filtered = []
    run = 0
    for i, a in enumerate(alerts):
        run = run + 1 if i>0 and alerts[i-1][1] == a[0]-1 else 1
        if run >= min_votes:
            filtered.append(a)
    print(f"[ALERT] {len(filtered)} triggers (thr={thr}, min_votes={min_votes})")
    return filtered

# 사용예:
# triggers = trigger_alert_on_video(df.iloc[0]["video_path"], model)
# print(triggers[:5])
```

```python
# ============================================================
# Sliding-window BiLSTM (강화판)
# - Pose (x,y,vis) → 정규화(중심/스케일) + 속도(Δ) 특징 추가
# - 윈도우 길이/보폭 튜닝: WIN=48, STRIDE=8
# - 백본: Bidirectional LSTM x2
# - 클래스 불균형 보정: class_weight
# - 임계값 자동 튜닝: 검증 F1 최대화 threshold
# - 결과: 혼동행렬/리포트 + 오분류 CSV 저장
# ============================================================

import os, cv2, numpy as np, pandas as pd, matplotlib.pyplot as plt
from tqdm import tqdm
import mediapipe as mp

# -----------------------------
# 0) 설정
# -----------------------------
DATASET_CSV = "./video_dataset_index.csv"   # ★ 네 경로로 수정
FPS_OUT = 20
WIN = 48                 # ← 길이 늘림
STRIDE = 8               # ← 보폭 줄임(촘촘)
MIN_VIS = 0.35

EPOCHS = 30
BATCH  = 64
INIT_THR = 0.5           # 초기 임계(최종 평가는 검증 기반 최적 thr 사용)

CSV_ALL = "eval_windows_all.csv"
CSV_ERR = "eval_windows_errors.csv"

# -----------------------------
# 1) MediaPipe Pose
# -----------------------------
mp_pose = mp.solutions.pose
POSE = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)
IDX = dict(L_HIP=23, R_HIP=24, L_SHOULDER=11, R_SHOULDER=12, NOSE=0)

def _vis_ok(lm, min_vis=MIN_VIS):
    pairs = [(IDX['L_SHOULDER'], IDX['R_SHOULDER']),
             (IDX['L_HIP'], IDX['R_HIP'])]
    for a, b in pairs:
        if lm[[a, b], 2].min() < min_vis:
            return False
    if lm[IDX['NOSE'], 2] < min_vis:
        return False
    return True

def frame_to_xyv(frame_bgr):
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    res = POSE.process(rgb)
    out = np.zeros((33, 3), dtype=np.float32)
    if res.pose_landmarks:
        for i, p in enumerate(res.pose_landmarks.landmark):
            out[i, 0] = p.x; out[i, 1] = p.y; out[i, 2] = p.visibility
    return out

# -----------------------------
# 2) 비디오 → (T,99) + 정규화
# -----------------------------
def video_to_seq99(video_path, fps_out=FPS_OUT):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return np.zeros((0, 99), np.float32)
    fps_in = cap.get(cv2.CAP_PROP_FPS) or fps_out
    step   = max(1, int(round(fps_in / float(fps_out))))
    feats, idx = [], 0
    while True:
        ok = cap.grab()
        if not ok: break
        if idx % step == 0:
            ok, frame = cap.retrieve()
            if not ok: break
            lm = frame_to_xyv(frame)
            if not _vis_ok(lm, MIN_VIS): lm[:] = 0.0
            feats.append(lm.reshape(-1))  # (99,)
        idx += 1
    cap.release()
    seq = np.asarray(feats, dtype=np.float32)  # (T,99)
    return normalize_seq99(seq)                #  정규화 적용

def normalize_seq99(seq):  # seq: (T,99)
    """엉덩이 중심으로 평행이동 + 몸통길이로 스케일 정규화."""
    if len(seq) == 0: return seq
    pts = seq.reshape(len(seq), 33, 3).copy()
    hip = (pts[:, IDX['L_HIP'], :2] + pts[:, IDX['R_HIP'], :2]) / 2.0         # (T,2)
    sh  = (pts[:, IDX['L_SHOULDER'], :2] + pts[:, IDX['R_SHOULDER'], :2]) / 2.0
    pts[:, :, 0] -= hip[:, 0:1]
    pts[:, :, 1] -= hip[:, 1:2]
    torso = np.linalg.norm(sh - hip, axis=1, keepdims=True) + 1e-6            # (T,1)
    pts[:, :, 0:2] /= torso[:, None, :]
    return pts.reshape(len(seq), -1).astype(np.float32)                        # (T,99)

# -----------------------------
# 3) 슬라이딩 윈도우 + 속도(Δ) 특징
# -----------------------------
def make_windows(seq99, win=WIN, stride=STRIDE, label=0):
    T = len(seq99)
    if T < win:
        zX = np.zeros((0, win, 99), np.float32)
        zy = np.zeros((0,), np.int64)
        zs = np.zeros((0,), np.int64)
        return zX, zy, zs
    Xs, ys, starts = [], [], []
    for s in range(0, T - win + 1, stride):
        e = s + win
        Xs.append(seq99[s:e])
        ys.append(label)
        starts.append(s)
    X = np.stack(Xs, axis=0).astype(np.float32)  # (N,win,99)
    X = add_vel_feature(X)                        # ★ 속도(Δ) 추가 → (N,win,198)
    y = np.asarray(ys, dtype=np.int64)
    starts = np.asarray(starts, dtype=np.int64)
    return X, y, starts

def add_vel_feature(X):  # X: (N, T, 99) → (N, T, 198)
    d = np.diff(X, axis=1, prepend=X[:, :1, :])
    return np.concatenate([X, d], axis=2)

# -----------------------------
# 4) 데이터셋 구성(비디오 단위 split)
# -----------------------------
from sklearn.model_selection import train_test_split

def build_dataset(df, fps_out=FPS_OUT, win=WIN, stride=STRIDE):
    if not np.issubdtype(df["label"].dtype, np.integer):
        df["label"] = df["label"].map({"non_fall":0, "Non-Fall":0, "fall":1, "Fall":1}).astype(int)

    tr_df, te_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])

    def build_from_df(dfx):
        X_list, y_list, paths, starts = [], [], [], []
        for _, row in tqdm(dfx.iterrows(), total=len(dfx), desc="Extract", unit="vid"):
            seq = video_to_seq99(row["video_path"], fps_out=fps_out)  # (T,99) normalized
            Xw, yw, sw = make_windows(seq, win=win, stride=stride, label=int(row["label"]))
            if len(Xw)==0: continue
            X_list.append(Xw); y_list.append(yw)
            paths  += [row["video_path"]]*len(sw)
            starts += [*sw]
        if not X_list:
            return (np.zeros((0, win, 198), np.float32),
                    np.zeros((0,), np.int64),
                    np.array([], dtype=object), np.array([], dtype=np.int64))
        X = np.concatenate(X_list, axis=0)
        y = np.concatenate(y_list, axis=0)
        paths  = np.array(paths,  dtype=object)
        starts = np.array(starts, dtype=np.int64)
        return X, y, paths, starts

    X_tr, y_tr, p_tr, s_tr = build_from_df(tr_df)
    X_te, y_te, p_te, s_te = build_from_df(te_df)
    info = dict(FPS_OUT=fps_out, WIN=win, STRIDE=stride)
    return (X_tr, y_tr, p_tr, s_tr), (X_te, y_te, p_te, s_te), info

# -----------------------------
# 5) 모델(BiLSTM) + 클래스가중치 + 임계값 튜닝
# -----------------------------
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay, classification_report

def build_model(input_shape, hid=128, drop=0.5):
    model = Sequential([
        Bidirectional(LSTM(hid, return_sequences=True), input_shape=input_shape),
        Dropout(drop),
        Bidirectional(LSTM(hid//2)),
        Dropout(drop),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

def best_threshold_by_f1(y_true, probs, lo=0.05, hi=0.95, steps=181):
    ts = np.linspace(lo, hi, steps)
    f1s = [f1_score(y_true, (probs>=t).astype(int)) for t in ts]
    idx = int(np.argmax(f1s))
    return float(ts[idx]), float(f1s[idx])

# -----------------------------
# 6) 실행
# -----------------------------
df = pd.read_csv(DATASET_CSV, encoding="utf-8-sig")
print(f"[INDEX] videos={len(df)}  pos={df['label'].sum()}  neg={len(df)-df['label'].sum()}")

(TR) , (TE), info = build_dataset(df, fps_out=FPS_OUT, win=WIN, stride=STRIDE)
X_tr, y_tr, path_tr, start_tr = TR
X_te, y_te, path_te, start_te = TE
print(f"[DATA] Train: X={X_tr.shape}, y={y_tr.shape} | Test: X={X_te.shape}, y={y_te.shape}")

# Train/Val 분리(윈도우 단위, stratify)
X_trn, X_val, y_trn, y_val = train_test_split(X_tr, y_tr, test_size=0.2, random_state=42, stratify=y_tr)

# 클래스 가중치 계산
w = compute_class_weight(class_weight='balanced', classes=np.array([0,1]), y=y_trn)
class_weight = {0: float(w[0]), 1: float(w[1])}
print("class_weight:", class_weight)

# 모델
model = build_model(input_shape=(X_tr.shape[1], X_tr.shape[2]), hid=128, drop=0.5)

cbs = [
    EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", patience=2, factor=0.5, verbose=1)
]

history = model.fit(
    X_trn, y_trn,
    epochs=EPOCHS, batch_size=BATCH, shuffle=True,
    validation_data=(X_val, y_val),
    callbacks=cbs, verbose=1,
    class_weight=class_weight     # ★ 불균형 보정
)

# 검증셋에서 최적 임계값 찾기
probs_val = model.predict(X_val, verbose=0).ravel()
thr_best, f1_best = best_threshold_by_f1(y_val, probs_val, lo=0.1, hi=0.9, steps=161)
print(f"[VAL] best_thr={thr_best:.3f}  best_f1={f1_best:.4f}")

# 테스트 평가(최적 thr 사용)
probs_te = model.predict(X_te, verbose=0).ravel()
y_pred  = (probs_te >= thr_best).astype(int)

try: auc = roc_auc_score(y_te, probs_te)
except: auc = float("nan")
print(f"[TEST] acc={accuracy_score(y_te, y_pred):.4f}  f1={f1_score(y_te, y_pred):.4f}  auc={auc:.4f}")

print("\n[Classification Report]\n" + classification_report(y_te, y_pred, target_names=["Non-Fall","Fall"]))

cm = confusion_matrix(y_te.astype(int), y_pred.astype(int), labels=[0,1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non-Fall","Fall"])
disp.plot(cmap="Blues", values_format='d')
plt.title("Confusion Matrix (BiLSTM + Norm+Vel + tuned thr)")
plt.tight_layout(); plt.show()

# 오분류 CSV 저장
df_all = pd.DataFrame({
    "video_path": path_te,
    "start_frame_resampled": start_te,
    "start_time_sec": start_te / float(info["FPS_OUT"]),
    "end_time_sec": (start_te + info["WIN"]) / float(info["FPS_OUT"]),
    "y_true": y_te.astype(int),
    "y_pred": y_pred.astype(int),
    "prob_fall": probs_te.astype(float),
})

err_df = df_all[df_all["y_true"] != df_all["y_pred"]].reset_index(drop=True)
df_all.to_csv(CSV_ALL, index=False, encoding="utf-8-sig")
err_df.to_csv(CSV_ERR, index=False, encoding="utf-8-sig")
print(f"[Saved] {CSV_ALL} ({len(df_all)})  |  {CSV_ERR} ({len(err_df)})")
```

```python
# ===========================
# A) 실험 설정(초 단위)
# ===========================
# fps_out: 출력 프레임레이트(프레임 간 시간 간격 = 1/fps_out 초)
# win_sec: 윈도우 길이(초), stride_sec: 보폭(초)
FPS_GRID     = [12, 15, 20, 24, 30]          # 초당 사용할 프레임
WIN_SEC_GRID = [2.0, 2.4, 2.8, 3.2]          # 창 길이(초)
STR_SEC_GRID = [0.4, 0.6, 0.8, 1.0]          # 보폭(초)  ← 겹침 조절

# 현재 최선 근접(20, 2.4, 0.8) 주변을 더 촘촘히 보고 싶다면:
# WIN_SEC_GRID = [2.2, 2.4, 2.6, 2.8]
# STR_SEC_GRID = [0.6, 0.8, 1.0]

# 조합 생성: stride < win 조건만 통과
from itertools import product
EXPERIMENTS = [
    {"fps_out": f, "win_sec": w, "stride_sec": s}
    for f, w, s in product(FPS_GRID, WIN_SEC_GRID, STR_SEC_GRID)
    if s < w
]

PATH = './fps-win-str_change_test_model/'
CSV_ALL = f"{PATH}/eval_windows_all.csv"
CSV_ERR = f"{PATH}/eval_windows_errors.csv"


# 간단한 포즈 캐시: 같은 fps_out 재사용(비디오 재처리 방지)
_POSE_SEQ_CACHE = {}  # key: (video_path, fps_out) -> np.ndarray(T,99)

# video_to_seq99를 캐시 버전으로 래핑
def video_to_seq99_cached(video_path, fps_out=FPS_OUT):
    key = (video_path, int(fps_out))
    if key in _POSE_SEQ_CACHE:
        return _POSE_SEQ_CACHE[key]
    seq = video_to_seq99(video_path, fps_out=fps_out)
    _POSE_SEQ_CACHE[key] = seq
    return seq

# build_dataset에서 video_to_seq99 호출 부분만 캐시 버전으로 교체
def build_dataset_cached(df, fps_out=FPS_OUT, win=WIN, stride=STRIDE):
    if not np.issubdtype(df["label"].dtype, np.integer):
        df["label"] = df["label"].map({"non_fall":0, "Non-Fall":0, "fall":1, "Fall":1}).astype(int)
    tr_df, te_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])

    def build_from_df(dfx):
        X_list, y_list, paths, starts = [], [], [], []
        for _, row in tqdm(dfx.iterrows(), total=len(dfx), desc="Extract", unit="vid"):
            seq = video_to_seq99_cached(row["video_path"], fps_out=fps_out)  # ← 캐시 사용
            Xw, yw, sw = make_windows(seq, win=win, stride=stride, label=int(row["label"]))
            if len(Xw)==0: continue
            X_list.append(Xw); y_list.append(yw)
            paths  += [row["video_path"]]*len(sw)
            starts += [*sw]
        if not X_list:
            return (np.zeros((0, win, 198), np.float32),
                    np.zeros((0,), np.int64),
                    np.array([], dtype=object), np.array([], dtype=np.int64))
        X = np.concatenate(X_list, axis=0)
        y = np.concatenate(y_list, axis=0)
        paths  = np.array(paths,  dtype=object)
        starts = np.array(starts, dtype=np.int64)
        return X, y, paths, starts

    X_tr, y_tr, p_tr, s_tr = build_from_df(tr_df)
    X_te, y_te, p_te, s_te = build_from_df(te_df)
    info = dict(FPS_OUT=fps_out, WIN=win, STRIDE=stride)
    return (X_tr, y_tr, p_tr, s_tr), (X_te, y_te, p_te, s_te), info

# ===========================
# B) 단일 실험 실행 함수
# ===========================
def run_experiment(df, fps_out, win_sec, stride_sec,
                   hid=128, drop=0.5, epochs=EPOCHS, batch=BATCH, init_thr=INIT_THR,
                   exp_tag=None):
    win    = max(2, int(round(win_sec   * fps_out)))    # 최소 길이 2 프레임 보장
    stride = max(1, int(round(stride_sec * fps_out)))

    (TR), (TE), info = build_dataset_cached(df, fps_out=fps_out, win=win, stride=stride)
    X_tr, y_tr, path_tr, start_tr = TR
    X_te, y_te, path_te, start_te = TE
    if len(X_tr)==0 or len(X_te)==0:
        return dict(exp_tag=exp_tag, fps_out=fps_out, win_sec=win_sec, stride_sec=stride_sec,
                    n_train=len(X_tr), n_test=len(X_te), acc=np.nan, f1=np.nan, auc=np.nan,
                    thr=np.nan)

    # Train/Val split
    X_trn, X_val, y_trn, y_val = train_test_split(X_tr, y_tr, test_size=0.2, random_state=42, stratify=y_tr)

    # class weight
    w = compute_class_weight(class_weight='balanced', classes=np.array([0,1]), y=y_trn)
    class_weight = {0: float(w[0]), 1: float(w[1])}

    # model
    model = build_model(input_shape=(X_tr.shape[1], X_tr.shape[2]), hid=hid, drop=drop)
    cbs = [
        EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True),
        ReduceLROnPlateau(monitor="val_loss", patience=2, factor=0.5, verbose=1)
    ]

    history = model.fit(
        X_trn, y_trn,
        epochs=epochs, batch_size=batch, shuffle=True,
        validation_data=(X_val, y_val),
        callbacks=cbs, verbose=0,
        class_weight=class_weight
    )

    # threshold tuning on val
    probs_val = model.predict(X_val, verbose=0).ravel()
    thr_best, f1_best = best_threshold_by_f1(y_val, probs_val, lo=0.1, hi=0.9, steps=161)

    # test
    probs_te = model.predict(X_te, verbose=0).ravel()
    y_pred   = (probs_te >= thr_best).astype(int)
    try: auc = roc_auc_score(y_te, probs_te)
    except: auc = float("nan")
    acc = accuracy_score(y_te, y_pred)
    f1  = f1_score(y_te, y_pred)

    # 결과 저장 파일명 접미사
    tag = exp_tag or f"fps{fps_out}_w{win}_s{stride}"
    csv_all = f"{CSV_ALL.replace('.csv','')}_{tag}.csv"
    csv_err = f"{CSV_ERR.replace('.csv','')}_{tag}.csv"
    model_path = f"model_{tag}.keras"

    # 혼동행렬 그림도 각 실험별로 저장
    cm = confusion_matrix(y_te.astype(int), y_pred.astype(int), labels=[0,1])
    fig, ax = plt.subplots()
    ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non-Fall","Fall"]).plot(
        cmap="Blues", values_format='d', ax=ax
    )
    ax.set_title(f"CM BiLSTM | {tag} | thr={thr_best:.2f}")
    plt.tight_layout()
    fig.savefig(f"cm_{tag}.png", dpi=150)
    plt.close(fig)

    # 전체/오분류 CSV
    df_all = pd.DataFrame({
        "video_path": path_te,
        "start_frame_resampled": start_te,
        "start_time_sec": start_te / float(info["FPS_OUT"]),
        "end_time_sec": (start_te + info["WIN"]) / float(info["FPS_OUT"]),
        "y_true": y_te.astype(int),
        "y_pred": y_pred.astype(int),
        "prob_fall": probs_te.astype(float),
    })
    err_df = df_all[df_all["y_true"] != df_all["y_pred"]].reset_index(drop=True)
    df_all.to_csv(csv_all, index=False, encoding="utf-8-sig")
    err_df.to_csv(csv_err, index=False, encoding="utf-8-sig")

    # 모델 가중치 저장
    model.save(model_path)

    return dict(
        exp_tag=tag, fps_out=fps_out, win_sec=win_sec, stride_sec=stride_sec,
        win_frames=win, stride_frames=stride,
        n_train=len(X_tr), n_val=len(X_val), n_test=len(X_te),
        acc=float(acc), f1=float(f1), auc=float(auc),
        thr=float(thr_best), f1_val=float(f1_best),
        csv_all=csv_all, csv_err=csv_err, cm_png=f"cm_{tag}.png", model_path=model_path
    )

# ===========================
# C) 전체 실험 실행 및 요약 테이블 저장
# ===========================
df = pd.read_csv(DATASET_CSV, encoding="utf-8-sig")
print(f"[INDEX] videos={len(df)}  pos={df['label'].sum()}  neg={len(df)-df['label'].sum()}")

summary = []
for i, conf in enumerate(EXPERIMENTS, 1):
    tag = f"exp{i:02d}"
    print(f"\n[RUN] {tag} :: fps_out={conf['fps_out']}, win={conf['win_sec']}s, stride={conf['stride_sec']}s")
    res = run_experiment(df, exp_tag=tag, **conf)
    summary.append(res)

summary_df = pd.DataFrame(summary)
summary_df = summary_df.sort_values(["f1", "auc", "acc"], ascending=[False, False, False]).reset_index(drop=True)
summary_df.to_csv("experiment_summary_timegrid.csv", index=False, encoding="utf-8-sig")
print("\n[SUMMARY Top 5]")
print(summary_df.head(5)[["exp_tag","fps_out","win_sec","stride_sec","f1","auc","acc","thr"]])
```

```python
# ============================================================
# Sliding-window LSTM for Fall/Non-Fall (runnable, improved)
# - Frame → (33,3) → (T,99) + velocity Δ → (T,165)
# - Sliding windows: (N, WIN, FEAT), label weakly from video label
# - Split by video to avoid leakage
# - Metrics, confusion matrix, error CSV
# ============================================================

import os, cv2, numpy as np, pandas as pd, matplotlib.pyplot as plt
from tqdm import tqdm
import mediapipe as mp

from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, f1_score, roc_auc_score,
                             confusion_matrix, ConfusionMatrixDisplay, classification_report)

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# -----------------------------
# 0) 설정
# -----------------------------
DATASET_CSV = "./video_dataset_index.csv"      # columns: [video_path, unique_name, label]
FPS_OUT = 24
WIN    = int(round(3.2 * FPS_OUT))
STRIDE = int(round(0.4 * FPS_OUT))
MIN_VIS = 0.35
BATCH  = 32
EPOCHS = 0            # 기존 모델 재학습 원치 않으면 0
THR    = 0.5          # 초기 임계 (이후 튜닝으로 갱신)
MODEL_PATH = "./fps-win-str_change_test_model/model_exp61.keras"

CSV_ALL = "eval_windows_all.csv"
CSV_ERR = "eval_windows_errors.csv"

# -----------------------------
# 1) MediaPipe Pose
# -----------------------------
mp_pose = mp.solutions.pose
POSE = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)
IDX = dict(L_HIP=23, R_HIP=24, L_SHOULDER=11, R_SHOULDER=12, NOSE=0)

def key_visibility_ok(lm, min_vis=MIN_VIS):
    pairs = [(IDX['L_SHOULDER'], IDX['R_SHOULDER']),
             (IDX['L_HIP'], IDX['R_HIP'])]
    for a, b in pairs:
        if lm[[a, b], 2].min() < min_vis:
            return False
    return lm[IDX['NOSE'], 2] >= min_vis

def frame_to_xyv(frame_bgr):
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    res = POSE.process(rgb)
    out = np.zeros((33, 3), dtype=np.float32)
    if res.pose_landmarks:
        for i, lm in enumerate(res.pose_landmarks.landmark):
            out[i, 0] = lm.x; out[i, 1] = lm.y; out[i, 2] = lm.visibility
    return out

# -----------------------------
# 2) Video → (T,165)  [99 + Δ66]
# -----------------------------
def video_to_sequence(video_path, fps_out=FPS_OUT):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return np.zeros((0, 165), np.float32)

    fps_in = cap.get(cv2.CAP_PROP_FPS)
    fps_in = fps_in if fps_in and fps_in > 0 else fps_out
    step = max(1, int(round(fps_in / float(fps_out))))

    feats, idx = [], 0
    while True:
        ok = cap.grab()
        if not ok: break
        if idx % step == 0:
            ok, frame = cap.retrieve()
            if not ok: break
            lm = frame_to_xyv(frame)  # (33,3)
            if not key_visibility_ok(lm, MIN_VIS):
                lm[:] = 0.0
            feats.append(lm.reshape(-1))  # 99
        idx += 1
    cap.release()

    if len(feats) == 0:
        return np.zeros((0, 165), np.float32)

    seq = np.asarray(feats, dtype=np.float32)  # (T,99)

    # velocity features
    xy  = seq[:, :66]
    dxy = np.diff(xy, axis=0, prepend=xy[:1, :])
    seq = np.concatenate([seq, dxy], axis=1).astype(np.float32)  # (T,165)

    return seq

# -----------------------------
# 3) Sliding windows
# -----------------------------
def make_windows(seq_2d, win=WIN, stride=STRIDE, label=0):
    T = len(seq_2d)
    F = 0 if T == 0 else seq_2d.shape[1]
    if T < win:
        return (np.zeros((0, win, F), np.float32),
                np.zeros((0,), np.int64),
                np.zeros((0,), np.int64))
    chunks, labs, starts = [], [], []
    for s in range(0, T - win + 1, stride):
        e = s + win
        chunks.append(seq_2d[s:e])
        labs.append(label)
        starts.append(s)
    X = np.stack(chunks, axis=0).astype(np.float32)
    y = np.asarray(labs, dtype=np.int64)
    starts = np.asarray(starts, dtype=np.int64)
    return X, y, starts

# -----------------------------
# 4) Build dataset (split by video)
# -----------------------------
def build_window_dataset(df, fps_out=FPS_OUT, win=WIN, stride=STRIDE):
    tr_df, te_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["label"])

    def build_from_df(dfx):
        X_list, y_list, meta_path, meta_start = [], [], [], []
        for _, row in tqdm(dfx.iterrows(), total=len(dfx), desc="Extract", unit="vid"):
            seq = video_to_sequence(row["video_path"], fps_out=fps_out)  # (T,165)
            Xw, yw, sw = make_windows(seq, win=win, stride=stride, label=int(row["label"]))
            if len(Xw)==0: 
                continue
            X_list.append(Xw); y_list.append(yw)
            meta_path += [row["video_path"]]*len(sw)
            meta_start.append(sw)

        if not X_list:
            return (np.zeros((0, win, 165), np.float32), np.zeros((0,), np.int64),
                    np.array([], dtype=object), np.array([], dtype=np.int64))

        X = np.concatenate(X_list, axis=0)
        y = np.concatenate(y_list, axis=0)
        meta_start = np.concatenate(meta_start, axis=0)
        meta_path = np.array(meta_path, dtype=object)

        # Optional undersampling to mitigate imbalance (simple, reproducible)
        pos_idx = np.where(y == 1)[0]
        neg_idx = np.where(y == 0)[0]
        if len(pos_idx) > 0 and len(neg_idx) > 0:
            k = min(len(pos_idx), len(neg_idx))
            rng = np.random.default_rng(42)
            pos_sel = rng.choice(pos_idx, size=k, replace=False)
            neg_sel = rng.choice(neg_idx, size=k, replace=False)
            keep = np.sort(np.concatenate([pos_sel, neg_sel]))
            X, y = X[keep], y[keep]
            meta_start = meta_start[keep]
            meta_path  = meta_path[keep]

        return X, y, meta_path, meta_start

    X_tr, y_tr, path_tr, start_tr = build_from_df(tr_df)
    X_te, y_te, path_te, start_te = build_from_df(te_df)
    info = dict(FPS_OUT=fps_out, WIN=win, STRIDE=stride)
    return (X_tr, y_tr, path_tr, start_tr), (X_te, y_te, path_te, start_te), info

# -----------------------------
# 5) Model
# -----------------------------
def build_model(input_shape, hid=128, drop=0.5):
    model = Sequential([
        LSTM(hid, input_shape=input_shape, return_sequences=True),
        Dropout(drop),
        LSTM(hid//2),
        Dropout(drop),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

def _normalize_input_shape(m):
    shp = getattr(m, "input_shape", None)
    if shp is None:
        return None
    if isinstance(shp, (list, tuple)) and len(shp) > 0 and isinstance(shp[0], (list, tuple)):
        shp = shp[0]
    return shp

def _need_rebuild(m, win_needed, feat_needed):
    shp = _normalize_input_shape(m)
    if shp is None: return True
    if not (isinstance(shp, (list, tuple)) and len(shp) == 3): return True
    _, t, f = shp
    return not (t == win_needed and f == feat_needed)

# -----------------------------
# 6) 실행
# -----------------------------
df = pd.read_csv(DATASET_CSV)
if not np.issubdtype(df["label"].dtype, np.integer):
    df["label"] = df["label"].map({"non_fall":0, "Non-Fall":0, "fall":1, "Fall":1}).astype(int)

print(f"[INDEX] videos={len(df)}  pos={df['label'].sum()}  neg={len(df)-df['label'].sum()}")

(TR) , (TE), info = build_window_dataset(df, fps_out=FPS_OUT, win=WIN, stride=STRIDE)
X_tr, y_tr, path_tr, start_tr = TR
X_te, y_te, path_te, start_te = TE
print(f"[DATA] Train: X={X_tr.shape}, y={y_tr.shape} | Test: X={X_te.shape}, y={y_te.shape}")

if len(X_tr)==0 and len(X_te)==0:
    raise RuntimeError("윈도우가 비어 있습니다. FPS_OUT/WIN/STRIDE 또는 비디오 경로를 확인해 주세요.")

FEAT    = int((X_tr.shape[2] if len(X_tr)>0 else X_te.shape[2]))
WIN_ACT = int((X_tr.shape[1] if len(X_tr)>0 else X_te.shape[1]))

# class weights
pos = int(y_tr.sum()); neg = int(len(y_tr) - pos) if len(y_tr)>0 else 0
pos_w = (neg / max(pos, 1)) if len(y_tr)>0 else 1.0
class_weight = {0: 1.0, 1: float(pos_w)}

# load or build
if os.path.exists(MODEL_PATH):
    model = load_model(MODEL_PATH)
    shp = _normalize_input_shape(model)
    print(f"[MODEL] loaded: {MODEL_PATH} | input_shape={shp}")
    if _need_rebuild(model, WIN_ACT, FEAT):
        print(f"[MODEL] 재구성: 요구 입력 (None,{WIN_ACT},{FEAT})")
        model = build_model(input_shape=(WIN_ACT, FEAT), hid=128, drop=0.5)
else:
    print("[MODEL] file not found → building new model")
    model = build_model(input_shape=(WIN_ACT, FEAT), hid=128, drop=0.5)

# train (optional)
if EPOCHS > 0 and len(X_tr)>0:
    import tensorflow as tf
    def f1_metric(y_true, y_pred):
        y_pred = tf.cast(y_pred >= 0.5, tf.float32)
        tp = tf.reduce_sum(y_true*y_pred)
        fp = tf.reduce_sum((1-y_true)*y_pred)
        fn = tf.reduce_sum(y_true*(1-y_pred))
        precision = tp / (tp + fp + 1e-7)
        recall    = tp / (tp + fn + 1e-7)
        return 2*precision*recall/(precision+recall+1e-7)

    model.compile(loss='binary_crossentropy', optimizer='adam',
                  metrics=['accuracy', tf.keras.metrics.AUC(name="auc"), f1_metric])

    cb = [
        EarlyStopping(monitor="val_f1_metric", mode="max", patience=6, restore_best_weights=True),
        ReduceLROnPlateau(monitor="val_f1_metric", mode="max", patience=3, factor=0.5, verbose=1),
    ]
    model.fit(
        X_tr, y_tr, epochs=EPOCHS, batch_size=BATCH,
        validation_split=0.2, shuffle=True, callbacks=cb, verbose=1,
        class_weight=class_weight
    )
    model.save(MODEL_PATH)
    print(f"[MODEL] saved: {MODEL_PATH}")

# threshold tuning on small val subset from train
if len(X_tr) > 0:
    X_sub, _, y_sub, _ = train_test_split(X_tr, y_tr, test_size=0.9, random_state=7, stratify=y_tr)
    if len(X_sub)>0:
        pv = model.predict(X_sub, verbose=0).ravel()
        thrs = np.linspace(0.1, 0.9, 33)
        best_thr, best_f1 = 0.5, -1
        for t in thrs:
            f1 = f1_score(y_sub, (pv>=t).astype(int))
            if f1 > best_f1:
                best_f1, best_thr = f1, float(t)
        THR = best_thr
        print(f"[THR] tuned threshold = {THR:.3f} (val_f1={best_f1:.3f})")

# -----------------------------
# 7) 평가/리포트
# -----------------------------
assert X_te.ndim==3 and X_te.shape[1]==WIN_ACT and X_te.shape[2]==FEAT, f"Bad X_te shape: {X_te.shape}"

loss, acc = model.evaluate(X_te, y_te, verbose=0)
print(f"[TEST] loss={loss:.4f}  acc={acc:.4f}")

probs = model.predict(X_te, verbose=0).ravel()
y_pred = (probs >= THR).astype(int)

try: auc = roc_auc_score(y_te, probs)
except: auc = float("nan")
print(f"[TEST] acc={accuracy_score(y_te, y_pred):.4f}  f1={f1_score(y_te, y_pred):.4f}  auc={auc:.4f}")
print("\n[Classification Report]\n" + classification_report(y_te, y_pred, target_names=["Non-Fall","Fall"]))

cm = confusion_matrix(y_te, y_pred, labels=[0,1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non-Fall", "Fall"])
disp.plot(cmap="Blues", values_format='d')
plt.title("Confusion Matrix (Sliding-window LSTM)")
plt.tight_layout(); plt.show()

df_all = pd.DataFrame({
    "video_path": path_te,
    "start_frame_resampled": start_te,
    "start_time_sec": start_te / float(FPS_OUT),
    "end_time_sec": (start_te + WIN_ACT) / float(FPS_OUT),
    "y_true": y_te.astype(int),
    "y_pred": y_pred.astype(int),
    "prob_fall": probs.astype(float),
})
err_df = df_all[df_all["y_true"] != df_all["y_pred"]].reset_index(drop=True)
df_all.to_csv(CSV_ALL, index=False, encoding="utf-8-sig")
err_df.to_csv(CSV_ERR, index=False, encoding="utf-8-sig")
print(f"[Saved] {CSV_ALL} ({len(df_all)})  |  {CSV_ERR} ({len(err_df)})")

# -----------------------------
# 8) (선택) 알림 트리거
# -----------------------------
def trigger_alert_on_video(video_path, model, fps_out=FPS_OUT, win=WIN_ACT, stride=1, thr=THR, min_votes=3):
    seq = video_to_sequence(video_path, fps_out=fps_out)  # (T,FEAT=165)
    if len(seq) < win:
        print("[ALERT] too short"); return []
    alerts, buf = [], []
    for t in range(len(seq)):
        buf.append(seq[t])
        if len(buf) > win: buf.pop(0)
        if len(buf) == win and (t % stride == 0):
            x = np.asarray(buf, dtype=np.float32)[None, ...]  # (1,win,FEAT)
            p = float(model.predict(x, verbose=0).ravel()[0])
            if p >= thr:
                alerts.append((t - win + 1, t, p))
    filtered, run = [], 0
    for i, a in enumerate(alerts):
        run = run + 1 if i>0 and alerts[i-1][1] == a[0]-1 else 1
        if run >= min_votes:
            filtered.append(a)
    print(f"[ALERT] {len(filtered)} triggers (thr={thr}, min_votes={min_votes})")
    return filtered
```

```python
import os, cv2, numpy as np
from dataclasses import dataclass
from tqdm import tqdm
import mediapipe as mp
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
from tensorflow import keras
import json

@dataclass
class DS:
    DATA_ROOT = r"./"                         # Kaggle 루트
    OUT_NPZ  = "imvia_xyv_win.npz"            # xyv 전용 npz
    TOP_DIRS = ("Coffee_room_01","Coffee_room_02","Home_01","Home_02","Lecture_room","Office")
    LABEL_MAP_JSON = ""                       # 선택: {"video_stem":"fall"|"normal", ...}

@dataclass
class BuildCfg:
    fps_out = 24                              # exp61과 동일
    win     = 77                              # 3.2s @24fps ≈ 76.8 → 77
    stride  = 10                              # 0.4s @24fps ≈ 9.6 → 10
    min_vis = 0.35
    save_npz = True

mp_pose = mp.solutions.pose
POSE = mp_pose.Pose(model_complexity=1, min_detection_confidence=0.5, min_tracking_confidence=0.5)
IDX_NOSE = 0

def get_landmarks_xyv(frame_bgr):
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    res = POSE.process(rgb)
    out = np.zeros((33, 3), dtype=np.float32)
    if res.pose_landmarks:
        for i, lm in enumerate(res.pose_landmarks.landmark):
            out[i, 0] = lm.x; out[i, 1] = lm.y; out[i, 2] = lm.visibility
    return out

def key_visibility_ok(lm, min_vis=0.35):
    return (lm[:,2].mean() >= min_vis) and (lm[IDX_NOSE,2] >= min_vis)


# path ----------

def _iter_scene_dirs(root, top_dirs):
    for d in top_dirs:
        p = os.path.join(root, d)
        if os.path.isdir(p): yield d, p

def _video_base_dir(scene_dir):
    direct = os.path.join(scene_dir, "Videos")
    if os.path.isdir(direct): return direct
    try:
        for name in os.listdir(scene_dir):
            sub = os.path.join(scene_dir, name)
            cand = os.path.join(sub, "Videos")
            if os.path.isdir(cand): return cand
    except: pass
    return scene_dir

def _iter_avi_recursive(base_dir):
    for r, _, files in os.walk(base_dir):
        for f in files:
            if f.lower().endswith(".avi"): yield os.path.join(r, f)

def infer_label_from_path(video_path: str) -> int:
    s = video_path.lower()
    if "non" in s or "normal" in s or "no_fall" in s: return 0
    if "fall" in s or "slip" in s or "collapse" in s: return 1
    return 0

def load_label_map(json_path: str):
    if not json_path: return {}
    try:
        with open(json_path, "r", encoding="utf-8") as f:
            raw = json.load(f)
        return {k:str(v).lower() for k,v in raw.items()}
    except: return {}

def label_from_map_or_path(video_path: str, label_map: dict) -> int:
    stem = os.path.splitext(os.path.basename(video_path))[0]
    if stem in label_map:
        v = label_map[stem]
        if v in ("1","fall","y","true"):  return 1
        if v in ("0","normal","n","false"): return 0
    return infer_label_from_path(video_path)
#-----------------------------------

def sample_video_to_windows(video_path, fps_out, win, stride, min_vis, video_label):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64)
    fps_in = cap.get(cv2.CAP_PROP_FPS)
    step = max(1, int(round(fps_in / float(fps_out)))) if fps_in and fps_in > 0 else 1

    frames = []
    idx = 0
    while True:
        ok = cap.grab()
        if not ok: break
        if idx % step == 0:
            ok, frame = cap.retrieve()
            if not ok: break
            lm = get_landmarks_xyv(frame)
            if not key_visibility_ok(lm, min_vis):
                frames.append(np.zeros((33,3), np.float32))
            else:
                frames.append(lm)
        idx += 1
    cap.release()

    if len(frames) < win:
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64)

    seq = np.stack(frames, axis=0).astype(np.float32)       # (T,33,3)
    X_t = seq.reshape(len(seq), 33*3)                       # (T,99)

    chunks, wlabs = [], []
    for s in range(0, len(X_t) - win + 1, stride):
        e = s + win
        chunks.append(X_t[s:e])
        wlabs.append(video_label)                            # 창 라벨 = 비디오 라벨

    X = np.stack(chunks, axis=0) if chunks else np.zeros((0, win, 99), np.float32)
    y = np.asarray(wlabs, dtype=np.int64) if wlabs else np.zeros((0,), np.int64)
    return X, y

def build_npz_from_kaggle(ds: DS, b: BuildCfg):
    lbl_map = load_label_map(ds.LABEL_MAP_JSON)
    X_list, y_list, g_list = [], [], []
    for scene, scene_dir in _iter_scene_dirs(ds.DATA_ROOT, ds.TOP_DIRS):
        base = _video_base_dir(scene_dir)       # 각 상위 폴더에서 실제 영상이 있는 Videos 위치를 자동 감지
        vids = list(_iter_avi_recursive(base))  # 하위 폴더까지 재귀 탐색해 .avi 파일 전체를 수집
        print(f"[SCAN] {scene}: {len(vids)}개 .avi")
        for v in tqdm(vids, desc=scene, unit="vid"):
            vid_lbl = label_from_map_or_path(v, lbl_map)
            Xv, yv = sample_video_to_windows(v, b.fps_out, b.win, b.stride, b.min_vis, vid_lbl) # 하나의 영상에서 (Nv, win, 99) 윈도우와 (Nv,) 창 라벨을 생성
            if len(Xv) == 0: continue       # 너무 짧거나 실패한 영상은 건너뜀
            X_list.append(Xv); y_list.append(yv)    # 누적 버퍼에 추가
            g_list += [os.path.splitext(os.path.basename(v))[0]] * len(yv)
    if not X_list: raise RuntimeError("윈도우가 만들어지지 않았습니다.")    # 한 개도 윈도우가 없으면 즉시 중단해 원인 파악
    X = np.concatenate(X_list, axis=0)  # 모든 영상을 첫 축(N)으로 이어 붙여 최종 (N, win, 99)를 만듦듦
    y = np.concatenate(y_list, axis=0)  # 라벨도 (N,)로 이어 붙임임
    groups = np.array(g_list)           # 그룹 ID를 넘파이 배열로 변환
    print(f"[NPZ] X={X.shape}, y={y.shape}, groups={groups.shape}")

    
    if b.save_npz:
        np.savez_compressed(ds.OUT_NPZ, X=X, y=y, groups=groups, info=dict(WIN=b.win, STRIDE=b.stride, FEAT="xyv"))
        print(f"저장 완료 → {ds.OUT_NPZ}")  # 저장 옵션이 켜져 있으면 NPZ로 압축 저장
    return X, y, groups



def _ensure_float32(X):
    return X.astype(np.float32, copy=False)

def _to_198_with_delta(X_win_99):
    # X_win_99: (N, T, 99)
    N, T, F = X_win_99.shape
    X3 = X_win_99.reshape(N, T, 33, 3)
    dX3 = np.diff(X3, axis=1, prepend=X3[:, :1])  # Δ(시간차), 첫 프레임은 0으로 유지
    X_cat = np.concatenate([X3, dX3], axis=-1)    # (N,T,33,6) = 198
    return X_cat.reshape(N, T, 198)


# 모델이 고정 길이를 요구해서 
def _align_timesteps(X, T_expected, pad_value=0.0):
    if X.ndim == 2:     # 배치 차원과 피처 차원만 있는 경우 시간 차원이 없다고 판단
        X = X[:, None, :]

    N, T, F = X.shape   # 배치 개수 N, 타임스텝 T, 피처 수 F를 분해
    if (T_expected is None) or (T == T_expected): return X      # 기대 길이가 미지정이거나 이미 일치하면 그대로 반환

    if T > T_expected:                  # 실제 길이가 더 길 때
        s = (T - T_expected) // 2       # 중앙 정렬을 위해 시작 오프셋을 계산
        return X[:, s:s+T_expected, :]  # 중앙 구간을 잘라 반환
    
    out = np.full((N, T_expected, F), pad_value, dtype=X.dtype) # 실제 길이가 더 짧을 때 zero padding
    out[:, :T, :] = X       # 앞쪽부터 원본을 복사합니다. 뒤쪽 나머지는 패딩 값으로 유지
    return out

def evaluate_with_keras_from_npz(npz_path: str, keras_model_path: str, thr: float = 0.375, batch: int = 512):
    npz = np.load(npz_path, allow_pickle=True)
    X, y, groups = npz["X"], npz["y"], npz["groups"]

    # 그룹 기반으로 학습·테스트 인덱스를 한 번 생성
    tr_idx, te_idx = next(GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42).split(X, y, groups))
    X_te, y_te = X[te_idx], y[te_idx] # 테스트 셋

    model = keras.models.load_model(keras_model_path)   # Keras 저장 모델을 메모리로 로드
    in_shape = model.input_shape if not isinstance(model.input_shape, list) else model.input_shape[0]
    _, T_expected, F_expected = in_shape  # e.g. (None, 77, 198)

    X_te = _ensure_float32(X_te)
    # 실제 길이가 다르면 중앙 크롭 또는 패딩으로 길이를 맞춤
    X_te = _align_timesteps(X_te, T_expected if T_expected is not None else X_te.shape[1])  

    # --- 핵심: 모델이 198을 기대하면 99→198 변환(Δ 붙이기) ---
    # 모델이 198피처를 기대하면 99피처 입력에 시간차 Δ를 붙여 198피처로 확장
    if F_expected == 198 and X_te.shape[-1] == 99:
        X_te = _to_198_with_delta(X_te)

    elif (F_expected is not None) and (X_te.shape[-1] != F_expected):
        X_te = X_te[..., :F_expected]

    print("[PRED] input ->", X_te.shape, "| model expects", model.input_shape)

    # 예측 확률을 얻고 1차원으로 평탄화
    probs = model.predict(X_te, batch_size=batch).reshape(-1)
    acc = accuracy_score(y_te, probs >= thr)
    f1  = f1_score(y_te, probs >= thr)

    # AUC가 계산 불가인 경우를 대비해 NaN을 설정
    try: auc = roc_auc_score(y_te, probs)
    except: auc = float("nan")
    print(f"[Keras {keras_model_path}] acc={acc:.4f} f1={f1:.4f} auc={auc:.4f} thr={thr:.3f}")

    # 혼동행렬 출력/저장
    y_pred = (probs >= thr).astype(np.int64)
    cm = confusion_matrix(y_te, y_pred); cmn = confusion_matrix(y_te, y_pred, normalize='true')
    print("\n[Confusion Matrix]\n", cm)
    print("\n[Classification Report]\n", classification_report(y_te, y_pred, target_names=["UPRIGHT(0)","FALL(1)"]))
    fig, ax = plt.subplots(figsize=(4.5,4.5)); im=ax.imshow(cm, cmap='Blues')
    ax.set_title("Confusion Matrix (exp61)"); ax.set_xlabel("Pred"); ax.set_ylabel("True")
    ax.set_xticks([0,1]); ax.set_yticks([0,1]); ax.set_xticklabels(["UPRIGHT(0)","FALL(1)"]); ax.set_yticklabels(["UPRIGHT(0)","FALL(1)"])
    for i in range(2):
        for j in range(2):
            ax.text(j,i,f"{cm[i,j]}\n({cmn[i,j]*100:.1f}%)",ha='center',va='center',fontsize=10)
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04); plt.tight_layout()
    plt.savefig("confusion_matrix_exp61.png", dpi=150, bbox_inches="tight"); plt.show()
    return acc, f1, auc


if __name__ == "__main__":
    ds = DS                  # 필요시 변경
    b  = BuildCfg()
    build_npz_from_kaggle(ds, b)
    evaluate_with_keras_from_npz(ds.OUT_NPZ, keras_model_path="./fps-win-str_change_test_model/model_exp61.keras", thr=0.375)
```

```python
# ============================================================
# MediaPipe Pose → 휴리스틱 라벨(프레임) → 슬라이딩 윈도우 → LSTM 분류
# - Kaggle IMVIA 폴더: 루트 바로 아래 6개 상위 폴더(스크린샷 구조 대응)
# - 샘플링: grab()/retrieve() 사용, fps_out으로 리샘플
# - 특징: (x,y,visibility) 33점 ⇒ 99차원/프레임 (0~1 범위 그대로)
# - 라벨: 휴리스틱 상태머신 (UPRIGHT=0 / FALLING,ON_GROUND=1)
# ============================================================
# pip install opencv-python mediapipe==0.10.15 numpy torch scikit-learn tqdm


# 중복 처리 => 재귀탐색, 경로 탐색
# 라벨 리라벨링 npz 숫자를 신뢰하지않고 경로기반 재매핑


# 슬라이딩 윈도우  이진 분류 
# 슬라이딩 윈도우 : 고정 길이의 프레임 묶음(윈도우)

# 몸의 각도(어깨,엉덩이의 벡터)를 통해 낙상을 구분하는 라벨을 만들고 슬라이딩 윈도우로 이진 분류  

import os, math, cv2, numpy as np
from dataclasses import dataclass
from typing import Tuple, List
from tqdm import tqdm
import mediapipe as mp
import torch, torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

# -----------------------------
# 0) 설정 (여기만 채워서 쓰세요)
# -----------------------------
@dataclass
class DS:
    # Kaggle 루트(스크린샷의 "낙상kaggle" 폴더)
    DATA_ROOT= r"./"      # 예: r"C:\Users\YUN\Documents\...\낙상kaggle"
    OUT_NPZ= "imvia_mppose_win.npz"
    # 루트 아래 상위 폴더들(존재하는 것만 사용)
    TOP_DIRS= ("Coffee_room_01","Coffee_room_02","Home_01","Home_02","Lecture_room","Office")


# LSTM 학습을 할 거면, 고정 길이 텐서가 필요해서 fps_out / win / stride 같은 세팅이 필수
@dataclass
class BuildCfg:
    fps_out= 24                              # 출력 샘플링 FPS (grab/retrieve로 리샘플)
    win = int(round(3.2 * fps_out))          # 윈도우 길이(프레임)
    stride = int(round(0.4 * fps_out))       # 윈도우 보폭(프레임)
    min_vis = 0.35                           # 가시성 품질 체크 임계
    save_npz = True                          # NPZ 저장 여부


@dataclass
class FallParams:         
    tilt_thr = 60.0
    hard_tilt = 65.0
    vel_fast_thr = 0.20
    dtilt_thr = 12.0
    height_thr = 0.85
    drop_thr = 0.20
    falling_max_sec = 2.0
    onground_hold_sec = 0.6

@dataclass
class TrainCfg:
    npz_path = DS.OUT_NPZ
    batch = 64
    epochs= 15
    lr = 1e-3
    hid = 128
    dropout = 0.2
    device = "cuda" if torch.cuda.is_available() else "cpu"

# -----------------------------
# 1) MediaPipe Pose
# -----------------------------
mp_pose = mp.solutions.pose
POSE = mp_pose.Pose(
    model_complexity=1,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

# Mediapipe 인덱스
IDX = dict(L_HIP=23, R_HIP=24, L_SHOULDER=11, R_SHOULDER=12, NOSE=0)

def get_landmarks_xyv(frame_bgr):
    """(33,3) = x,y,visibility (0~1). 실패 시 0."""
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    res = POSE.process(rgb)
    out = np.zeros((33, 3), dtype=np.float32)
    if res.pose_landmarks:
        for i, lm in enumerate(res.pose_landmarks.landmark):
            out[i, 0] = lm.x
            out[i, 1] = lm.y
            out[i, 2] = lm.visibility
    return out

def key_visibility_ok(lm, min_vis=0.35):
    pairs = [(IDX['L_SHOULDER'], IDX['R_SHOULDER']),
             (IDX['L_HIP'], IDX['R_HIP'])]
    for a, b in pairs:
        if lm[[a, b], 2].min() < min_vis:
            return False
    if lm[IDX['NOSE'], 2] < min_vis:
        return False
    return True

def _center_xy(landmarks):
    """어깨/엉덩이 중심 기준 평행이동(형상 각도 계산용)"""
    pts = landmarks[:, :2].copy()
    lhip, rhip = pts[IDX['L_HIP']], pts[IDX['R_HIP']]
    center = (lhip + rhip) / 2
    pts -= center
    return pts

def torso_tilt_deg(pts_centered):
    lsh, rsh = pts_centered[IDX['L_SHOULDER']], pts_centered[IDX['R_SHOULDER']]
    lhip, rhip = pts_centered[IDX['L_HIP']], pts_centered[IDX['R_HIP']]
    sh_c = (lsh + rsh) / 2
    hip_c = (lhip + rhip) / 2
    v = sh_c - hip_c
    ang = abs(math.degrees(math.atan2(v[0], -v[1])))  # 수직 기준; 서있음≈0°, 누울수록↑
    return float(ang)

def frame_label_heuristic(lm,
                          last_hip_y_raw,
                          fps_out,
                          params: FallParams,
                          last_tilt_s):
    """
    프레임 라벨링을 위해 필요한 값들 계산 + 상태 업데이트에 쓸 신호 반환.
    반환: (label(0/1), tilt_deg, dtilt_dps, head_h, hip_h, vel)
    """
    # (형상) 기울기
    pts_c = _center_xy(lm)
    tilt = torso_tilt_deg(pts_c)

    # (절대) 높이/속도
    head_h_raw = 1.0 - lm[IDX['NOSE'], 1]
    hip_y_raw = 0.5 * (lm[IDX['L_HIP'], 1] + lm[IDX['R_HIP'], 1])
    hip_h_raw = 1.0 - hip_y_raw

    # 엉덩이의 수직 속도. Δy × fps_out → 양수면 아래로 급강하.
    vel = 0.0 if last_hip_y_raw is None else (hip_y_raw - last_hip_y_raw) * fps_out

    # 간단 스무딩: 지수이동평균 한 스텝만(부담 줄이기)
    if last_tilt_s is None:
        tilt_s = tilt
    else:
        SMOOTH_WIN = 5
        k = 2 / (SMOOTH_WIN + 1)  # smooth_win=5 고정
        tilt_s = last_tilt_s * (1 - k) + tilt * k
    dtilt_s = 0.0 if last_tilt_s is None else (tilt_s - last_tilt_s) * fps_out

    # 임계 기반 라벨(프레임 단위; 간단 버전)

    # **‘자세가 낮다’**는 신호. 다음 중 하나라도 참이면 True
    # 기울기가 충분히 큼(tilt_s ≥ tilt_thr) // 머리 높이/엉덩이 높이가 낮음(≤ height_thr) // 머리 낙폭이 큼(1 - head_h_raw ≥ drop_thr)
    low_posture = ( (tilt_s >= params.tilt_thr) or (head_h_raw <= params.height_thr)
                    or (hip_h_raw <= params.height_thr) or ((1.0 - head_h_raw) >= params.drop_thr) )
    
    # 급격한 변화 신호:
    # 엉덩이의 하강 속도가 빠르다. 
    # 기울기 변화속도가 빠르고(dtilt_s > dtilt_thr) 현재 기울기도 어느 정도 있다(tilt_s > 30) 
    # 하드 기울기(매우 누움) 자체로도 급변으로 인정.
    fast_drop  = (vel > params.vel_fast_thr) or (dtilt_s > params.dtilt_thr and tilt_s > 30) or (tilt_s >= params.hard_tilt)

    label = 1 if (low_posture and fast_drop) or tilt_s >= params.hard_tilt else 0

    return label, tilt_s, dtilt_s, head_h_raw, hip_h_raw, vel


# -----------------------------
# 2) 비디오 스캔 및 시퀀스 생성
# -----------------------------
def _iter_scene_dirs(root ,top_dirs):
    for d in top_dirs:
        p = os.path.join(root, d)
        if os.path.isdir(p):    # 실제로 폴더가 존재할 때만
            yield d, p          # ("Home_01", "/root/Home_01") 같은 튜플을 생성

def _video_base_dir(scene_dir):
    """scene/Videos 또는 scene/*/Videos 자동 인식"""
    direct = os.path.join(scene_dir, "Videos")
    if os.path.isdir(direct):
        return direct
    
    try:
        for name in os.listdir(scene_dir):          # scene의 하위 폴더들을 뒤져서
            sub = os.path.join(scene_dir, name)
            cand = os.path.join(sub, "Videos")
            if os.path.isdir(cand):
                return cand
    except Exception:
        pass
    
    return scene_dir


def _iter_avi_recursive(base_dir):

        # r은 현재 디렉터리 경로, _는 (사용하지 않는) 하위 폴더 목록, files는 파일 목록
    for r, _, files in os.walk(base_dir):       # os.walk() => base_dir 아래의 모든 하위 폴더를 재귀적으로 순회
        for f in files:
            if f.lower().endswith(".avi"):
                yield os.path.join(r, f)



def sample_video_to_windows(video_path,
                            fps_out,
                            win,
                            stride,
                            min_vis,
                            params: FallParams):
    """
    비디오 1개 → (N,win,99), (N,)
    - grab/retrieve로 fps_out 샘플링
    - 프레임 입력: (x,y,vis)*33 → 99
    - 프레임 라벨: 휴리스틱(0/1)
    - 윈도우 라벨: 창 안에 1이 존재하면 1
    """

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64)

    fps_in = cap.get(cv2.CAP_PROP_FPS) or fps_out
    step = max(1, int(round(fps_in / float(fps_out))))

    frames, labels = [], []
    idx = 0
    last_hip_y_raw = None
    last_tilt_s = None

    while True:
        ok = cap.grab()
        if not ok:
            break

        if idx % step == 0:
            ok, frame = cap.retrieve()
            if not ok:
                break

            lm = get_landmarks_xyv(frame)  # (33,3) 0~1 // 33개 관절의 (x, y, visibility)

            if not key_visibility_ok(lm, min_vis):
                # 시그널/특징 품질이 낮으면 제로
                frames.append(np.zeros((33,3), np.float32))
                labels.append(0)
            else:               # frame_label_heuristic가 기울기/변화속도/머리·엉덩이 높이/속도를 계산하고 프레임 라벨(0/1) 을 반환
                label, tilt_s, dtilt_s, head_h, hip_h, vel = frame_label_heuristic(
                    lm, last_hip_y_raw, fps_out, params, last_tilt_s
                )
                frames.append(lm)
                labels.append(label)
                last_hip_y_raw = 0.5 * (lm[IDX['L_HIP'], 1] + lm[IDX['R_HIP'], 1])
                last_tilt_s = tilt_s

        idx += 1    # 프레임 인덱스 증가 
    cap.release()
    

    if len(frames) < win:
        return np.zeros((0, win, 99), np.float32), np.zeros((0,), np.int64)

    seq = np.stack(frames, axis=0).astype(np.float32)     # (T,33,3)
    X_t = seq.reshape(len(seq), 33*3)                     # (T,99)
    y_t = np.asarray(labels, dtype=np.int64)              # (T,)

    # 슬라이딩 윈도우
    chunks, wlabs = [], []
    for s in range(0, len(X_t) - win + 1, stride):
        e = s + win
        chunks.append(X_t[s:e])
        wlabs.append(1 if y_t[s:e].max() == 1 else 0)

    X = np.stack(chunks, axis=0) if chunks else np.zeros((0, win, 99), np.float32)
    y = np.asarray(wlabs, dtype=np.int64) if wlabs else np.zeros((0,), np.int64)
    return X, y



def build_npz_from_kaggle(ds: DS, b: BuildCfg, fp: FallParams):
    # 필수 설정 확인. 루트 경로가 비어 있으면 즉시 중단
    root_use = ds.DATA_ROOT if ds.DATA_ROOT else os.getcwd()

    X_list, y_list, g_list = [], [], []

    for scene, scene_dir in _iter_scene_dirs(root_use, ds.TOP_DIRS):        # ("Home_01", "/root/Home_01") 같은 튜플
        base = _video_base_dir(scene_dir)   # video 폴더 
        vids = list(_iter_avi_recursive(base))  # .avi 동영상
        print(f"[SCAN] {scene}: {len(vids)}개 .avi")

        for v in tqdm(vids, desc=scene, unit="vid"):

            # Xv: (N_v, win, 99) (이 영상에서 나온 윈도우 수 × 윈도우 길이 × 99특징)
            # yv: (N_v,) (각 윈도우의 0/1 라벨)
            # 윈도우 : 연속된 프레임 win장을 붙여 만든 고정 길이 시퀀스
            # stride: 다음 윈도우를 만들 때 시작 프레임을 얼마나 건너뛸지
            Xv, yv = sample_video_to_windows(v, b.fps_out, b.win, b.stride, b.min_vis, fp)
            if len(Xv) == 0: 
                continue
            X_list.append(Xv)
            y_list.append(yv)
            g_list += [os.path.splitext(os.path.basename(v))[0]] * len(yv)
            # os.path.basename(v)는 파일명 // splitext(...)[0]는 확장자를 뗀 영상 이름

    if not X_list:
        raise RuntimeError("윈도우가 만들어지지 않았습니다.")

    X = np.concatenate(X_list, axis=0)  # (N,win,99)
    y = np.concatenate(y_list, axis=0)  # (N,)
    groups = np.array(g_list)

    print(f"[NPZ] X={X.shape}, y={y.shape}, groups={groups.shape}")
    if b.save_npz:
        np.savez_compressed(ds.OUT_NPZ, X=X, y=y, groups=groups,
                            info=dict(WIN=b.win, STRIDE=b.stride, FEAT="xyv(33*3)"))
        print(f"저장 완료 → {ds.OUT_NPZ}")
    return X, y, groups


from sklearn.metrics import confusion_matrix, classification_report   
import matplotlib.pyplot as plt                                       


# eval_confusion_matrix
def eval_confusion_matrix(model,
                          te_dl,
                          y_true,
                          device,
                          thr= 0.5,
                          save_png = "confusion_matrix_best.png"):
    """
    베스트 모델로 테스트 세트 예측 → 컨퓨전 매트릭스/리포트 출력 및(옵션) 저장
    """
    model.eval()
    probs = []
    with torch.no_grad():
        for xb, _ in te_dl:
            xb = xb.to(device)
            p = torch.sigmoid(model(xb)).cpu().numpy()
            probs.append(p)
    probs = np.concatenate(probs)
    y_pred = (probs >= thr).astype(np.int64)

    # 1) 텍스트 출력
    cm = confusion_matrix(y_true, y_pred)
    print("\n[Confusion Matrix] (rows=true, cols=pred)")
    print(cm)
    print("\n[Classification Report]")
    print(classification_report(y_true, y_pred, target_names=["UPRIGHT(0)", "FALL(1)"]))

    # 2) 시각화 (카운트 + 비율 함께 표기)
    cm_norm = confusion_matrix(y_true, y_pred, normalize='true')
    fig, ax = plt.subplots(figsize=(4.5, 4.5))
    im = ax.imshow(cm, cmap='Blues')
    ax.set_title("Confusion Matrix (best F1)")
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")
    ax.set_xticks([0, 1]); ax.set_yticks([0, 1])
    ax.set_xticklabels(["UPRIGHT(0)", "FALL(1)"])
    ax.set_yticklabels(["UPRIGHT(0)", "FALL(1)"])

    # 셀 안에 count와 비율(%) 표기
    for i in range(2):
        for j in range(2):
            ax.text(j, i, f"{cm[i, j]}\n({cm_norm[i, j]*100:.1f}%)",
                    ha="center", va="center", fontsize=10)

    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    plt.tight_layout()
    if save_png:
        plt.savefig(save_png, dpi=150, bbox_inches="tight")
        print(f"[Saved] {save_png}")
    plt.show()

    # 필요 시 개별 지표 반환
    return {
        "cm": cm,
        "cm_norm": cm_norm,
        "y_pred": y_pred,
        "probs": probs,
    }


# -----------------------------
# 3) LSTM 학습/평가
# -----------------------------
class WinDS(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X).float()
        self.y = torch.from_numpy(y).float()
    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]


class LSTMCls(nn.Module):
    def __init__(self, feat_dim=99, hid=128, dropout=0.2):
        super().__init__()
        self.lstm = nn.LSTM(input_size=feat_dim, hidden_size=hid, num_layers=1,
                            batch_first=True, bidirectional=True)
        self.head = nn.Sequential(
            nn.LayerNorm(hid*2),
            nn.Dropout(dropout),
            nn.Linear(hid*2, hid),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hid, 1)
        )

        
    def forward(self, x):
        out, _ = self.lstm(x)          # (B,T,2H)
        logits = self.head(out[:, -1]) # 마지막 타임스텝
        return logits.squeeze(1)


def train_lstm_from_npz(cfg: TrainCfg):
    npz = np.load(cfg.npz_path, allow_pickle=True)
    X, y, groups = npz["X"], npz["y"], npz["groups"]
    print("[LOAD]", X.shape, y.shape, groups.shape)

    # 그룹 단위 셔플 분할기를 생성 //  한 번만 분할하며 검증 비율은 20%
    gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
    tr, te = next(gss.split(X, y, groups))
    X_tr, y_tr = X[tr], y[tr]
    X_te, y_te = X[te], y[te]

    pos = y_tr.sum(); neg = len(y_tr) - pos
    pos_w = neg / max(pos, 1)
    print(f"pos_w={pos_w:.2f} (pos={int(pos)}, neg={int(neg)})")

    tr_dl = DataLoader(WinDS(X_tr, y_tr), batch_size=cfg.batch, shuffle=True)
    te_dl = DataLoader(WinDS(X_te, y_te), batch_size=cfg.batch, shuffle=False)

    model = LSTMCls(feat_dim=X.shape[2], hid=cfg.hid, dropout=cfg.dropout).to(cfg.device)
    opt = torch.optim.AdamW(model.parameters(), lr=cfg.lr)
    crit = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([pos_w], device=cfg.device))

    best = (-1, None)
    for ep in range(1, cfg.epochs+1):
        model.train(); tr_loss = 0.0
        for xb, yb in tr_dl:    # (입력, 타깃) 반복
            xb, yb = xb.to(cfg.device), yb.to(cfg.device)
            opt.zero_grad()                     # 이전 스텝의 누적 기울기를 0으로 초기화
            logit = model(xb)                   # 순전파로 로짓(시그모이드 이전 점수)을 계산
            loss = crit(logit, yb)              # 예측 로짓과 타깃으로 손실을 계
            loss.backward(); opt.step()         # 역전파로 기울기를 계산하고, 옵티마이저로 파라미터를 한 스텝 갱신
            tr_loss += loss.item() * len(xb)    # 배치 손실을 샘플 수 가중 합으로 누적
        tr_loss /= len(tr_dl.dataset)           # 에포크 학습 손실의 샘플 단위 평균을 계산

        model.eval(); preds, probs = [], []
        with torch.no_grad():   # 파이토치의 자동미분을 비활성화하여 메모리·속도를 최적화
            for xb, _ in te_dl:
                xb = xb.to(cfg.device)
                p = torch.sigmoid(model(xb)).cpu().numpy()  # 시그모이드로 확률로 변환하고, CPU 넘파이 배열로 가져옴옴
                preds.append((p >= 0.5).astype(np.int64)); probs.append(p)  # 0.5 임계로 확률을 이진 라벨로 변환해 preds에, 원본 확률을 probs에 누적
        preds = np.concatenate(preds); probs = np.concatenate(probs)        
        acc = accuracy_score(y_te, preds); f1 = f1_score(y_te, preds)
        try: auc = roc_auc_score(y_te, probs)
        except: auc = float("nan")
        print(f"[Ep{ep}] loss={tr_loss:.4f} | acc={acc:.4f} f1={f1:.4f} auc={auc:.4f}")

        if f1 > best[0]:
            best = (f1, model.state_dict()) # F1이 최고치를 갱신하면 현재 모델 가중치를 베스트로 저장

        if best[1] is not None: 
            torch.save(best[1], "lstm_mppose_best.pt")
            print("베스트 저장 → lstm_mppose_best.pt")

            # === 추가: 베스트 모델 로드 후 컨퓨전 매트릭스/리포트 출력 ===
            
            model.load_state_dict(best[1])  # 메모리의 베스트 가중치를 모델에 로드
            
            _ = eval_confusion_matrix(model, te_dl, y_te, cfg.device, thr=0.5,
                                    save_png="confusion_matrix_best.png")
    

# -----------------------------
# 4) 실행
# -----------------------------
if __name__ == "__main__":
    ds = DS()              # ★여기에 Kaggle 루트 입력
    b  = BuildCfg()
    hp = FallParams()                     # 휴리스틱 임계치 (필요 시 미세조정)

    # (1) 윈도우 데이터 생성(+저장)
    X, y, groups = build_npz_from_kaggle(ds, b, hp)

    # (2) LSTM 학습/평가
    cfg = TrainCfg()                 # 키워드 인자 쓰지 말고 기본 생성
    cfg.npz_path = ds.OUT_NPZ        # 원하는 경로를 속성에 직접 대입
    train_lstm_from_npz(cfg)
```

```python
# ============================
# 5) 단일 영상 감지 + 오버레이 + (옵션) 이메일
# ============================


# 초기 설정 => naver 메일 홈페이지 들어가서 POP3/SMTP => 사용함으로 킴
# POP3/SMTP "사용함" 눌러놓고 혹시 모르니 보안설정하기 2차 인증 등등 
# 앱에서 앱 비밀번호 발급받기

# send_email_smtp 함수내에 메일과 "앱" 비밀번호 입력 ( 그냥 비밀번호는 안됨 )

# "injooha@naver.com"

EMAIL_TO = ("rrt-y@naver.com", )

import time, smtplib
from email.mime.text import MIMEText

def load_best_model(device="cpu", feat_dim=99, hid=128, dropout=0.2, ckpt="lstm_mppose_best.pt"):

    # 저장 당시와 동일한 하이퍼파라미터로 LSTMCls 인스턴스를 생성하고 곧바로 .to(device)로 CPU나 GPU로 이동
    m = LSTMCls(feat_dim=feat_dim, hid=hid, dropout=dropout).to(device)

    # 디스크의 체크포인트 파일을 읽어 파이썬 객체로 복원
    # map_location=device로 저장 당시와 다른 장치여도 현재 지정한 장치 메모리로 매핑
    sd = torch.load(ckpt, map_location=device)

    # 읽어온 가중치 딕셔너리를 모델 파라미터에 주입
    # 가 정확히 일치해야 하며, 기본적으로 strict=True라서 누락, 여분 키가 있으면 예외를 발생
    m.load_state_dict(sd)
    m.eval()
    return m

def send_email_smtp(subject, body,
                    smtp_host="smtp.naver.com", smtp_port=587,
                    sender="rrt-y@naver.com",
                    password="3596GUP96X5P",     # 네이버 앱 비밀번호 발급 후 입력 
                    to=EMAIL_TO):
    """
    naver 계정의 앱 비밀번호 필요. 위 sender/password/to만 채우면 됨.
    원치 않으면 이 함수 호출을 주석 처리하면 됨.
    """
    try:
        msg = MIMEText(body, "html", "utf-8")
        msg["Subject"] = subject
        msg["From"] = sender
        msg["To"] = ", ".join(to)

        # SMTP 세션 객체를 만들고, 포트 587로 평문 연결 후 STARTTLS로 TLS로 업그레이드 // 보안을 위해 필수
        s = smtplib.SMTP(smtp_host, smtp_port); s.starttls()

        # 송신자 주소와 앱 비밀번호로 인증하고, 실제 전송을 수행
        # sendmail의 두 번째 인자는 수신자 리스트이며, 세 번째 인자는 전체 MIME 메세지의 문자열 직렬화
        s.login(sender, password); s.sendmail(sender, list(to), msg.as_string())
        s.quit()    # 종료
    except Exception as e:
        print("[EMAIL] fail:", e)
        # 실패 시 원인 메시지를 [EMAIL] fail: 형식으로 출력


def run_video_with_overlay(video_path,
                           out_path="alert_overlay.mp4",
                           fps_out=None,          # None이면 입력 fps 유지
                           thr=0.5,
                           min_votes=3,           # 연속 판정 N회 이상일 때 '사건'으로 트리거
                           email_when_trigger=False,
                           email_to=EMAIL_TO,
                           device=None,
                           bcfg: BuildCfg=None,
                           hcfg: FallParams=None):
    """
    - MediaPipe로 프레임→(33,3)→99 특징, 슬라이딩 버퍼 길이=win
    - 버퍼가 꽉 찰 때마다 LSTM 확률 추정
    - 확률>=thr가 min_votes번 연속이면 'FALL DETECTED' 오버레이 표시하고 이메일(옵션) 발송
    """
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    bcfg = bcfg or BuildCfg()   # 전처리, 윈도우 길이 등 기본 빌드 설정이 없으면 새로 생성
    hcfg = hcfg or FallParams() # 휴리스틱 파라미터가 없으면 기본값으로 생성

    # 모델 로드
    # LSTM 분류기를 지정 장치에 로드하고 eval() 상태로 준비
    model = load_best_model(device=device, feat_dim=99, hid=128, dropout=0.2, ckpt="lstm_mppose_best.pt")

    cap = cv2.VideoCapture(video_path)   # 입력 비디오를 열기
    if not cap.isOpened():
        raise SystemExit(f"cannot open: {video_path}")
        # 열기에 실패하면 즉시 종료

    # fps를 읽고 실패시 지정 fps로 
    fps_in = cap.get(cv2.CAP_PROP_FPS) or bcfg.fps_out
    W = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))      # 프레임 가로 크기
    H = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))     # 프레임 세로 크기
    fps_write = fps_in if fps_out is None else fps_out      # 출력 FPS를 결정
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")        # MP4V 코덱을 네 글자 코드로 변환
    vw = cv2.VideoWriter(out_path, fourcc, fps_write, (W, H))   # 결과 영상을 기록할 비디오 라이터를 생성

    step = 1  # 입력 fps 그대로 처리(프레임 스킵 원하면 >1로)
    buf = []  # (win,99)
    votes = 0   # 연속으로 임계치 이상이 나온 횟수 카운터
    event_on = False    # 현재 경보 상태 플래그
    last_email_time = 0

    # 포즈 준비 (한 번만 생성)
    with mp.solutions.pose.Pose(model_complexity=1,
                                min_detection_confidence=0.5,
                                min_tracking_confidence=0.5) as pose:
        fidx = 0
        while True:
            ok = cap.grab()
            if not ok:
                break
            if fidx % step != 0:
                fidx += 1
                continue
            ok, frame = cap.retrieve()
            if not ok:
                break

            # 특징 추출
            # 프레임에서 33개 관절의 (x,y,visibility)를 얻습니다. shape (33,3)
            lm = get_landmarks_xyv(frame)
            # 가시도 조건이 나쁘면 99차원 영(0) 특징으로 대체
            if not key_visibility_ok(lm, bcfg.min_vis):
                feat = np.zeros((99,), np.float32)  
            else:
                # 여기선 휴리스틱 라벨은 쓰지 않고, 특징만 사용
                # 정상일 때 (33,3)을 평탄화해 (99,) 특징으로 만듦듦
                feat = lm.reshape(-1).astype(np.float32)

            # 최신 프레임 특징을 버퍼 끝에 추가
            buf.append(feat)
            # 길이가 win을 넘으면 가장 오래된 것을 제거
            if len(buf) > bcfg.win:
                buf.pop(0)

            # 현재 프레임의 낙상 확률을 저장할 변수 초기화
            prob = None
            # 버퍼가 꽉 찼을 때만 모델 추론을 수행
            if len(buf) == bcfg.win:
                # 버퍼를 (1, win, 99) 텐서로 만들어 장치로 올림 
                x = torch.from_numpy(np.asarray(buf)[None, ...]).float().to(device)  # (1,win,99)
                with torch.no_grad():
                    prob = torch.sigmoid(model(x)).item()   # 그래디언트 없이 추론하고 로짓을 시그모이드로 확률화해 파이썬 float으로 꺼냄

                if prob >= thr:
                    votes += 1  # 임계치 이상이면 연속 카운트를 증가
                else:
                    votes = 0   # 임계치 미만이면 연속 카운트를 리셋

            # 오버레이용 작업 복사본
            vis = frame.copy()

            # 현재 재생 시간을 초 단위
            tsec = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000.0

            # 기본 HUD 텍스트에 시간을 넣습니다.
            txt = f"t={tsec:6.2f}s" 

            # 추론한 경우 확률 표시
            if prob is not None:
                txt += f"  p={prob:0.3f}"

            # 좌하단에 시간·확률 텍스트를 흰색으로 그림
            cv2.putText(vis, txt, (12, H-12), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 2, cv2.LINE_AA)


            # 연속 임계치 충족 횟수가 기준 이상이면 이벤트로 간주
            if votes >= min_votes:
                # 이벤트 진입
                if not event_on:
                    event_on = True # 경보 상태로 전환
                    print(f"[ALERT] FALL DETECTED at {tsec:.2f}s  p={prob:.3f}")
                    if email_when_trigger and (time.time() - last_email_time > 30):
                        send_email_smtp(
                                        subject="[낙상 감지]",
                                        body=f"비디오: {os.path.basename(video_path)}\n"
                                            f"시간: {tsec:.2f}초\n"
                                            f"확률: {prob:.3f}",
                                        to=email_to
                                    )
                        last_email_time = time.time()
                        
                # 시각 경고 배너
                cv2.rectangle(vis, (0,0), (W, 50), (0,0,255), -1)
                cv2.putText(vis, "FALL DETECTED", (12, 35),
                            cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255,255,255), 3, cv2.LINE_AA)
            else:
                event_on = False

            vw.write(vis)
            fidx += 1

    cap.release(); vw.release()
    print(f"[SAVE] {out_path}")
```
