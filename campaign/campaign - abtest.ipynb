```python
import numpy as np
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import ttest_ind

# CSV 읽기(세미콜론 구분)
df_control = pd.read_csv("./control_group.csv", sep=";", engine="python", encoding="utf-8-sig")
df_test = pd.read_csv("./test_group.csv",    sep=";", engine="python", encoding="utf-8-sig")

cols = df_test.columns.str.replace(r'^\s*#\s*', '', regex=True)
df_control.columns = cols 
df_test.columns = cols 

# 합치기
df = pd.concat([df_control, df_test], ignore_index=True)

# 숫자형 변환 및 결측 0
num_cols = [
    "Spend [USD]",
    "of Impressions",
    "Reach",
    "of Website Clicks",
    "of Searches",
    "of View Content",
    "of Add to Cart",
    "of Purchase",
]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

# 그룹 라벨 생성
df["group"] = df["Campaign Name"].str.contains("Test", case=False).map({True: "test", False: "control"})

# 집계 및 전환율
# - 캠페인 이름 / 날짜 / 지출 / 노출 수 / 도달 범위 / 웹사이트 클릭 수 / 검색 수 / 콘텐츠 조회 수 / 장바구니 추가 수 / 구매 수 
agg = (
    df.groupby("group")[num_cols]
      .sum()
      .assign(
          ctr = lambda x: x["of Website Clicks"] / x["of Impressions"].replace(0, np.nan),     
          # 클릭률 ==> 웹사이트 클릭 수 / 노출 수 

          view_rate = lambda x: x["of View Content"] / x["of Website Clicks"].replace(0, np.nan), 
          # 상세페이지 조회율 ==>  콘텐츠 조회 수 / 웹사이트 클릭 수

          atc_rate = lambda x: x["of Add to Cart"] / x["of View Content"].replace(0, np.nan),     
          # 장바구니 전환율 ==> 장바구니 추가 수 / 콘텐츠 조회 수

          purchase_rate_click = lambda x: x["of Purchase"] / x["of Website Clicks"].replace(0, np.nan),   
          # 클릭 대비 구매율 ==> 구매 수 / 웹사이트 클릭 수

          purchase_rate_view = lambda x: x["of Purchase"] / x["of View Content"].replace(0, np.nan),     
          # 상세조회 대비 구매율 ==> 구매 수 / 콘텐츠 조회 수

          purchase_rate_imp = lambda x: x["of Purchase"] / x["of Impressions"].replace(0, np.nan),      
          # 노출 대비 구매율 ==> 구매 수 / 노출 수
      ) 
)

# A/B 비율검정 함수(전환율 차이, 리프트, z-test, 95% CI)
def ab_rate_test(df_agg, success_col, denom_col, alt="two-sided",
                 label_a="control", label_b="test"):
    a_succ = df_agg.loc[label_a, success_col]
    b_succ = df_agg.loc[label_b, success_col]
    a_den  = df_agg.loc[label_a, denom_col]
    b_den  = df_agg.loc[label_b, denom_col]

    z, p = proportions_ztest([a_succ, b_succ], [a_den, b_den], alternative=alt)

    pa = a_succ / a_den if a_den > 0 else np.nan
    pb = b_succ / b_den if b_den > 0 else np.nan
    diff = pb - pa
    se = np.sqrt( pa*(1-pa)/a_den + pb*(1-pb)/b_den ) 
    ci95 = diff - 1.96*se, diff + 1.96*se
    lift = pb/pa - 1

    return {
        "rate_control": pa, "rate_test": pb,
        "diff_test_minus_control": diff,
        "lift_test_over_control": lift,
        "z_stat": z, "p_value": p, "ci95_diff": ci95,
        "success_a": a_succ, "success_b": b_succ,
        "denom_a": a_den, "denom_b": b_den
    }


# 예시 검정: 구매/웹사이트클릭 전환율
res_purchase_click = ab_rate_test(
    agg, success_col="of Purchase", denom_col="of Website Clicks", alt="two-sided"
)
print("Purchase per Click:", res_purchase_click)

# 보조 지표 검정 예시
res_ctr   = ab_rate_test(agg, "of Website Clicks", "of Impressions")
res_view  = ab_rate_test(agg, "of View Content", "of Website Clicks")
res_atc   = ab_rate_test(agg, "of Add to Cart", "of View Content")
res_p_imp = ab_rate_test(agg, "of Purchase", "of Impressions")

print("CTR:", res_ctr)
print("ViewRate:", res_view)
print("ATC:", res_atc)
print("Purchase/Impr:", res_p_imp)

# 일자 단위 평균 전환율 Welch t-검정(선택)
daily = (
    df.assign(Date=pd.to_datetime(df["Date"], dayfirst=True))
      .groupby(["group", "Date"], as_index=False)
      .agg({"of Website Clicks":"sum", "of Impressions":"sum",
            "of View Content":"sum", "of Purchase":"sum"})
)
daily["purchase_rate_click"] = daily["of Purchase"] / daily["of Website Clicks"].replace(0, np.nan)

a = daily.loc[daily["group"]=="control", "purchase_rate_click"].dropna()
b = daily.loc[daily["group"]=="test",    "purchase_rate_click"].dropna()
t_stat, p_val = ttest_ind(a, b, equal_var=False, nan_policy="omit")
print({"daily_mean_control": a.mean(), "daily_mean_test": b.mean(),
       "t_stat": t_stat, "p_value": p_val})
```

```python
import mpmath as mp
mp.mp.dps = 80  # 정밀도 ↑

def p_from_z(z, two_sided=True):

    # erfc => 보정오차함수의 여함수 / 표준정규분포의 양측 꼬리확률과 동일한 값을 반환
    # erfc(|z|/√2) => 양쪽 꼬리 확률의 합
    p = mp.erfc(abs(mp.mpf(z))/mp.sqrt(2))    # 두측 p
    return p if two_sided else p/2

print('\nP_value\n')
for name, res in {"Purchase/Click":res_purchase_click,
                  "CTR":res_ctr, "ViewRate":res_view,
                  "ATC":res_atc, "Purchase/Impr":res_p_imp}.items():
    print(name, p_from_z(res["z_stat"]))
```

- 퍼널 분석

```python
import numpy as np
import pandas as pd

# 1) 단계와 분자/분모 정의
STAGES = [
    ("Impressions",      "of Impressions", None),                 # 시작점(분모 없음)
    ("Website Clicks",   "of Website Clicks",  "of Impressions"),
    ("View Content",     "of View Content",    "of Website Clicks"),
    ("Add to Cart",      "of Add to Cart",     "of View Content"),
    ("Purchase",         "of Purchase",        "of Add to Cart"),  
]
# 누적 전환율은 Impressions 대비 Purchase 등으로 별도 계산


# 2) 그룹 라벨이 없다면 생성
if "group" not in df.columns:
    df["group"] = df["Campaign Name"].str.contains("Test", case=False).map({True:"test", False:"control"})

# 3) 그룹별 합계 및 비용 지표
g = (df.groupby("group", as_index=True)[
            #  지출      /     노출 수     / 웹사이트 클릭 수    / 콘텐츠 조회 수 /   장바구니 추가 수  / 구매 수 
        ["Spend [USD]","of Impressions","of Website Clicks","of View Content","of Add to Cart","of Purchase"]
    ].sum())

#  Cost Per Mille 노출 1,000회당 지출 비용
g["CPM"] = g["Spend [USD]"] / g["of Impressions"] * 1000

# Cost Per Click 클릭 1회당 지출 비용
g["CPC"] = g["Spend [USD]"] / g["of Website Clicks"].replace(0, np.nan)

# Cost Per Acquisition/Action 전환 1건당 지출 비용
g["CPA"] = g["Spend [USD]"] / g["of Purchase"].replace(0, np.nan)

# 4) 퍼널 표 구성
rows = []
for name, succ, denom in STAGES:
    row = {"stage": name}
    # 수량
    row["count_control"] = g.loc["control", succ]
    row["count_test"] = g.loc["test", succ]
    # 단계 전환율(step rate)
    if denom is None:
        row["step_rate_control"] = np.nan
        row["step_rate_test"] = np.nan
    else:
        row["step_rate_control"] = g.loc["control", succ] / g.loc["control", denom] if g.loc["control", denom] else np.nan
        row["step_rate_test"] = g.loc["test", succ] / g.loc["test", denom] if g.loc["test", denom] else np.nan

    # 드롭오프율(= 1 - step_rate) // 어떤 퍼널 단계에서 다음 단계로 넘어가지 못하고 이탈한 비율
    # 드롭오프율 => 1 − 단계 전환율 = (현재 단계 수 − 다음 단계 수) ÷ 현재 단계 수
    # 퍼널 =>  노출 -> 웹 클릭 -> 콘텐츠 조회 -> 장바구니 -> 구매
    row["dropoff_control"] = 1 - row["step_rate_control"] if pd.notna(row["step_rate_control"]) else np.nan
    row["dropoff_test"] = 1 - row["step_rate_test"] if pd.notna(row["step_rate_test"]) else np.nan


    # 차이와 리프트
    if pd.notna(row.get("step_rate_control")) and pd.notna(row.get("step_rate_test")):

        # Control 전환율을 a, Test 전환율을 b
        a, b = row["step_rate_control"], row["step_rate_test"]
        row["diff_test_minus_control"] = b - a

        # 리프트(lift)는 테스트 전환율이 컨트롤 전환율 대비 상대적으로 얼마나 늘었거나 줄었는지를 나타내는 지표
        # 리프트 = (테스트 전환율 ÷ 컨트롤 전환율) − 1
        row["lift_test_over_control"]  = (b/a - 1) if (a and a>0) else np.nan

    rows.append(row)

funnel_df = (pd.DataFrame(rows)
               .set_index("stage")
               .reindex([s[0] for s in STAGES]))

# 5) 누적 전환율(노출 대비)
cum = {}
for grp in ["control","test"]:
    imp = g.loc[grp, "of Impressions"]
    cum[grp] = {
        "to_click": g.loc[grp,"of Website Clicks"]/imp if imp else np.nan,
        "to_view": g.loc[grp,"of View Content"]/imp if imp else np.nan,
        "to_cart": g.loc[grp,"of Add to Cart"]/imp if imp else np.nan,
        "to_purchase": g.loc[grp,"of Purchase"]/imp if imp else np.nan,
    }
cum_df = (pd.DataFrame(cum)
            .rename(index={"to_click":"cumulative: Impr→Click",
                           "to_view":"cumulative: Impr→View",
                           "to_cart":"cumulative: Impr→ATC",
                           "to_purchase":"cumulative: Impr→Purchase"}))

# 6) 비용 요약 함께 보기
cost_df = g[["Spend [USD]","CPM","CPC","CPA"]]

# funnel_df, cum_df, cost_df
```

```python
def outliers_percentage(data):

    outliers_perc = []

    for k,v in data.items():
        if data[k].dtype != 'O':
            q1 = v.quantile(0.25)
            q3 = v.quantile(0.75)
            irq = q3 - q1
            v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
            perc = np.shape(v_col)[0] * 100.0 / np.shape(data)[0]
            out_tuple = (k,int(perc))
            outliers_perc.append(out_tuple)
            print("Column %s outliers = %.2f%%" % (k,perc))

outliers_percentage(df_control)
```

```python
def remove_outliers_iqr(data, column):

    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter the data
    filtered_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return filtered_data


capped_control_group = df_control.copy()
capped_test_group = df_test.copy()

for col in features:

    capped_control_group = remove_outliers_iqr(capped_control_group, col)
    capped_test_group = remove_outliers_iqr(capped_test_group, col)
```

Shapiro test

정규성 검정을 위한 Shapiro–Wilk 검정은 정규성에서 벗어나는 모든 형태의 일탈을 탐지하도록 고안된 검정입니다. p-값이 0.05 이하이면 정규성 가설을 기각

```python
from scipy.stats import shapiro

for metric in metrics:

    print(f'\n{metric}:')

    test_stat, pvalue = shapiro(capped_control_group[metric])
    print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))

    test_stat, pvalue = shapiro(capped_test_group[metric])
    print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

Levene’s Test

Levene 검정은 n개의 표본이 동일한 분산을 가지는지를 확인하기 위한 검정<br>
 p값이 0.05 미만이면 등분산 가정이 위배되었다고 해석

```python
for metric in metrics:
    levene_test = stats.levene(capped_control_group[metric], capped_test_group[metric])
    print(f"\nLevene's Test for Homogeneity of Variances: {metric}")
    print(f"Statistic = {levene_test.statistic}, p-value = {levene_test.pvalue}")
```

- CTR 유의하다

---
Levene 검정 결과와 일치하는지 확인하기 위해 t-검정 

가설 검정

귀무가설(H0): 컨트롤 그룹과 테스트 그룹의 성과에는 유의한 차이가 없다.

대립가설(H1): 컨트롤 그룹과 테스트 그룹의 성과에는 유의한 차이가 있다.

```python
for metric in metrics:

    print(f'\n{metric}:')
    t_stat, p_value = stats.ttest_ind(capped_control_group[metric], capped_test_group[metric])
    print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

- CTR 유의
- 클릭률(CTR)을 제외하면, 95% 신뢰수준에서 대립가설을 기각하고 귀무가설을 채택

정규분포가 아닌 변수에 대한 검정(Mann–Whitney U 검정)

이는 독립표본 t-검정의 비모수적 대안으로, 정규성 가정을 요구하지 않으며 표본 수가 작을 때도 잘 작동

```python
for metric in metrics:

    print(f'\n{metric}:')
    print(stats.mannwhitneyu(capped_control_group[metric], capped_test_group[metric]))
```

- CTR만 유의
