```python
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])
orders["order_approved_at"] = pd.to_datetime(orders["order_approved_at"])
orders["order_delivered_carrier_date"] = pd.to_datetime(orders["order_delivered_carrier_date"])
orders["order_delivered_customer_date"] = pd.to_datetime(orders["order_delivered_customer_date"])
orders["order_estimated_delivery_date"] = pd.to_datetime(orders["order_estimated_delivery_date"])

reviews["review_creation_date"] = pd.to_datetime(reviews["review_creation_date"])
reviews["review_answer_timestamp"] = pd.to_datetime(reviews["review_answer_timestamp"])
```

---
주문 단위 집계 (매출, 운송비, 상품 수)

```python
order_items
```

```python
item_agg = order_items.groupby('order_id').agg(
                n_items = ('order_item_id','count'),
                products_price_sum = ('price','sum'),
                freight_value_sum = ('freight_value','sum')
            ).reset_index()

# 주문별 총 매출(상품가 + 운송비)
item_agg["order_gross_revenue"] = item_agg["products_price_sum"] + item_agg["freight_value_sum"]

item_agg
```

---
payments → 주문 단위 결제 정보 집계

```python
payments
```

```python
pay_agg = (
    payments
    .groupby("order_id")
    .agg(
        payment_total=("payment_value", "sum"),
        n_payments=("payment_sequential", "max")
    )
    .reset_index()
)
pay_agg
```

---
reviews → 주문 단위 리뷰 정보 집계

```python
rev_agg = (
    reviews
    .groupby("order_id")
    .agg(
        review_score=("review_score", "mean"),        # 주문에 리뷰가 여러 개면 평균
        first_review_date=("review_creation_date", "min")
    )
    .reset_index()
)
rev_agg
```

customers → 고객 지역 정보

```python
cust_sel = customers[[
    "customer_id",
    "customer_unique_id",
    "customer_city",
    "customer_state"
]]
cust_sel
```

Join

```python
# 고객 정보 붙이기
order_base = orders.merge(cust_sel, on="customer_id", how="left")

# 주문 아이템 집계 붙이기
order_base = order_base.merge(item_agg, on="order_id", how="left")

# 결제 정보 붙이기
order_base = order_base.merge(pay_agg, on="order_id", how="left")

# 리뷰 정보 붙이기
order_base = order_base.merge(rev_agg, on="order_id", how="left")

order_base.head()
```

```python
# 주문 월, 주문 요일
order_base["order_date"] = order_base["order_purchase_timestamp"].dt.date
order_base["order_month"] = order_base["order_purchase_timestamp"].dt.to_period("M")
order_base["order_dow"] = order_base["order_purchase_timestamp"].dt.dayofweek  # 0=월 ~ 6=일
order_base
```

```python
# 배송시간 ( 고객이 실제로 상품을 받은 날 - 고객이 주문한 시각 )
order_base["delivery_days"] = (
    order_base["order_delivered_customer_date"] - order_base["order_purchase_timestamp"]
).dt.days
order_base
```

```python
# 예상 대비 지연일수 ( 고객이 실제로 상품을 받은 날 - 시스템이 고객에게 약속한 예정 배송일 )
order_base["delay_vs_estimated"] = (
    order_base["order_delivered_customer_date"] - order_base["order_estimated_delivery_date"]   
).dt.days
order_base
```

```python
# 객단가, 주문당 평균 매출(AOV) : 주문별 결제 총액 / 상품 수
order_base["aov"] = order_base["payment_total"] / order_base["n_items"]
order_base
```

```python
def classify_delay(x):
    if pd.isna(x):
        return "unknown"
    if x > 0:
        return "delayed"      # 지연
    elif x < 0:
        return "early"        # 예상보다 빨리
    else:
        return "on_time"      # 정확히 예상일

shipping_df["delay_bucket"] = shipping_df["delay_vs_estimated"].apply(classify_delay)

delay_summary = (
    shipping_df["delay_bucket"]
    .value_counts(normalize=True)
    .rename("ratio")
    .to_frame()
)

delay_summary
```

```python
from googletrans import Translator
import pandas as pd
from tqdm import tqdm

translator = Translator()
tqdm.pandas()  # progress_apply용

def pt_to_ko(text):
    if not isinstance(text, str) or not text.strip():
        return ""
    try:
        res = translator.translate(text, src="pt", dest="ko")
        return res.text
    except Exception as e:
        # 에러 나면 일단 빈 문자열/None 반환
        return None
```

```python
target["review_comment_message_ko"] = target["review_comment_message"].progress_apply(pt_to_ko)
```

# 배송 관련 결론 

1) 일부 주(AL, AM, AP, RR)는<br>
전체 평균 대비 배송 리드타임이 크게 길고(25~30일 이상),<br>
고객 입장에서 체감하는 배송 품질이 가장 낮은 지역<br>
<br>
이는 지리적 거리, 인프라, 물류망 특성 등 지역 자체의 구조적인 요인이 클 가능성이 있다

- 해결 방안 : 물류 거점 추가, 배송 옵션 차등 안내, 예상 배송일 산정 로직 개선 등등

2) 주문 수 30건 이상인 셀러 중, 평균 배송일과 지연 비율이 모두 높은 ‘문제 셀러 그룹’을 정의하였고 <br>
이 문제 셀러들이 처리한 주문은 MG, RJ, SP, RS 주에 집중되어 있으며,<br>
평균 배송일(약 20~30일)과 지연 비율(약 15~31%)을 전체 평균보다 크게 끌어올리고 있다

- 개개인 셀러의 문제점을 면밀하게 파악하여 더 나은 방향성을 모색할 필요가 있다

```python
if "customer_unique_id" in order_base.columns:
    customer_key = "customer_unique_id"
else:
    customer_key = "customer_id"
customer_key
```

```python
order_base["delay_bucket"] = order_base["delay_vs_estimated"].apply(classify_delay)
```

```python
customer_master = (
    order_base
    .groupby(customer_key)
    .agg(
        n_orders=("order_id", "nunique"),                       # 주문 수
        first_order=("order_purchase_timestamp", "min"),        # 첫 주문일
        last_order=("order_purchase_timestamp", "max"),         # 마지막 주문일
        total_payment=("payment_total", "sum"),                 # 누적 결제 금액
        avg_aov=("aov", "mean"),                                # 평균 객단가
        avg_delivery_days=("delivery_days", "mean"),            # 평균 배송일
        delayed_ratio=("delay_bucket",
                       lambda s: (s == "delayed").mean()),      # 지연 비율
        avg_review_score=("review_score", "mean")               # 평균 리뷰 점수
    )
    .reset_index()
)

# 고객 생애기간 LTV (첫 주문~마지막 주문 일수)
customer_master["first_order"] = pd.to_datetime(customer_master["first_order"])
customer_master["last_order"]  = pd.to_datetime(customer_master["last_order"])

customer_master["customer_lifetime_days"] = (
    customer_master["last_order"] - customer_master["first_order"]
).dt.days

# 재구매 여부 (주문 2회 이상이면 1)
customer_master["is_repeat"] = (customer_master["n_orders"] > 1).astype(int)

customer_master.head()
```

```python
# 전체 고객 수
print("고객 수:", len(customer_master))

# 재구매율
repeat_rate = customer_master["is_repeat"].mean()
print("재구매율(2회 이상 주문 비율): {:.1%}".format(repeat_rate))

# 고객당 누적 매출 분포
display(customer_master["total_payment"].describe())

# 재구매 vs 1회성 고객 비교
display(
    customer_master
    .groupby("is_repeat")[["total_payment", "avg_aov",
                           "avg_delivery_days", "delayed_ratio",
                           "avg_review_score"]]
    .mean()
)
```

```python
import matplotlib.pyplot as plt

# 0/1을 라벨로 바꾸기
tmp = (
    customer_master
    .groupby("is_repeat")["total_payment"]
    .mean()
    .rename(index={0: "1회성 고객", 1: "재구매 고객"})
)

plt.figure(figsize=(5,4))
tmp.plot(kind="bar")
plt.title("1인당 누적 매출(총 결제금액) 비교")
plt.ylabel("평균 누적 결제 금액")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 기준 시점: 마지막 주문 다음 날
ref_date = order_base["order_purchase_timestamp"].max() + pd.Timedelta(days=1)

rfm = customer_master.copy()
rfm["recency"] = (ref_date - rfm["last_order"]).dt.days      # 마지막 주문이 얼마나 최근인가
rfm["frequency"] = rfm["n_orders"]                           # 주문 횟수
rfm["monetary"] = rfm["total_payment"]                       # 누적 매출

rfm[["recency","frequency","monetary"]].describe()
```

```python
# 낮을수록 좋은 recency는 qcut 순서를 반대로
rfm["R_score"] = pd.qcut(rfm["recency"], 3, labels=[3,2,1]).astype(int)
rfm["F_score"] = pd.qcut(rfm["frequency"].rank(method="first"), 3, labels=[1,2,3]).astype(int)
rfm["M_score"] = pd.qcut(rfm["monetary"].rank(method="first"), 3, labels=[1,2,3]).astype(int)

rfm["RFM_score"] = rfm["R_score"] + rfm["F_score"] + rfm["M_score"]

# 최고 가치 고객(8~9점)과 최저 가치 고객(3~4점) 비교
seg_summary = (
    rfm.assign(segment=np.where(rfm["RFM_score"] >= 8, "VIP",
                                np.where(rfm["RFM_score"] <= 4, "Low", "Mid")))
    .groupby("segment")[["monetary","frequency","recency"]]
    .mean()
)

seg_summary
```

```python
import pandas as pd
import numpy as np

# 0) 기준 날짜 설정 (마지막 주문 다음 날을 기준으로)
ref_date = order_base["order_purchase_timestamp"].max() + pd.Timedelta(days=1)

# 1) customer_master에 R/F/M 기본 값 만들기
customer_master["recency"]  = (ref_date - customer_master["last_order"]).dt.days
customer_master["frequency"] = customer_master["n_orders"]
customer_master["monetary"]  = customer_master["total_payment"]

# 2) 재구매 고객(3.1%)만 추출
repeat_cust = customer_master[customer_master["is_repeat"] == 1].copy()

print("재구매 고객 수 :", len(repeat_cust))

# 3) 재구매 고객들에 대해 RFM 점수(1~3점) 부여
# recency는 숫자가 작을수록 좋은 것이므로 qcut 라벨을 3,2,1 순서로 줍니다.
repeat_cust["R_score"] = pd.qcut(
    repeat_cust["recency"], 3, labels=[3, 2, 1]
).astype(int)

# frequency, monetary는 클수록 좋으니 1,2,3 순서
repeat_cust["F_score"] = pd.qcut(
    repeat_cust["frequency"].rank(method="first"), 3, labels=[1, 2, 3]
).astype(int)

repeat_cust["M_score"] = pd.qcut(
    repeat_cust["monetary"].rank(method="first"), 3, labels=[1, 2, 3]
).astype(int)

# 4) 최종 RFM 점수(3~9점)
repeat_cust["RFM_score"] = (
    repeat_cust["R_score"] + repeat_cust["F_score"] + repeat_cust["M_score"]
)

# 5) 재구매 고객 RFM 분포 빠르게 확인
print(
    repeat_cust[["recency","frequency","monetary","R_score","F_score","M_score","RFM_score"]]
    .describe()
)

# 6) 예시: VIP 후보(예: RFM_score >= 8)만 보기
vip_repeat = repeat_cust[repeat_cust["RFM_score"] >= 8].copy()
print("재구매 고객 중 VIP 후보 수:", len(vip_repeat))
vip_repeat.head()
```
